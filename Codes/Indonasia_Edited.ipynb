```python
import pandas as pd
import numpy as np


Iran_to_Indonasia=pd.read_excel(f"C:/TradeGroup/Reports/Iran-Indonasia-Be.xlsx")
Iran_to_Indonasia['شماره تعرفه'].fillna(0,inplace=True)
Iran_to_Indonasia['HS8']=Iran_to_Indonasia['شماره تعرفه'].astype(int)
Iran_Exports_Codes = pd.unique(Iran_to_Indonasia['HS8'])
Iran_Exports_Codes

My_Excel_File=pd.ExcelFile(f"C:/TradeGroup/Reports/Iran-Indonasia-Alef.xlsx")
dfs = []
for sheet_name in My_Excel_File.sheet_names:
    df= My_Excel_File.parse(sheet_name)
    dfs.append(df)

The_Commodities = pd.concat(dfs,ignore_index=True)
The_Commodities.rename(columns={'شماره تعرفه':"HS8"},inplace=True)
The_Commodities['HS8']=pd.to_numeric(The_Commodities['HS8'])
Iran_Imports_Codes = pd.unique(The_Commodities['HS8'])
Iran_Imports_Codes
```

```python
baseyear= 1395
for year in range(1395,1403):


   Exports = pd.read_excel(f'C:/TradeGroup/Data/Cleaned_Data/Y{year}Exports.xlsx')
   Exports_Indonasia = Exports[Exports['Country']=="اندونزي"]


   Imports = pd.read_excel(f'C:/TradeGroup/Data/Cleaned_Data/Y{year}Imports.xlsx')
   Imports_Indonasia = Imports[Imports['Country']=="اندونزي"]

   #Exports['Country'].unique()
   #Exports_Indonasia



   Exports_Indonasia['Tarjihi']=np.where(np.isin(Exports_Indonasia['Tcode'], Iran_Exports_Codes),1,0 )
   Exports_Indonasia_Tarjihi=Exports_Indonasia[Exports_Indonasia['Tarjihi']==1]
   #Exports_Indonasia_Tarjihi


   Imports_Indonasia['Tarjihi']=np.where(np.isin(Imports_Indonasia['Tcode'], Iran_Imports_Codes),1,0 )
   Imports_Indonasia_Tarjihi=Imports_Indonasia[Imports_Indonasia['Tarjihi']==1]
   #Imports_Indonasia_Tarjihi




   Exports_Idn=Exports_Indonasia.groupby('Tcode').agg(
               Tarjihi= ('Tarjihi','first'),
               Tdesc = ('Tdesc','first'),
               Dollars_Tot=('Dollars','sum')
               
               
            #   Tcode= ('Tcode','first')
         ).reset_index()

   Exports_Idn['Total_Idn_year']=Exports_Idn['Dollars_Tot'].sum()
   Exports_Idn['Total_Idn_Tarjihi_year']=Exports_Idn[Exports_Idn['Tarjihi']==1]['Dollars_Tot'].sum()




   Imports_Idn=Imports_Indonasia.groupby('Tcode').agg(
               Tarjihi= ('Tarjihi','first'),
               Tdesc = ('Tdesc','first'),
               Dollars_Tot=('Dollars','sum')
               
               
            #   Tcode= ('Tcode','first')
         ).reset_index()
   Imports_Idn['Total_Idn_year']=Imports_Idn['Dollars_Tot'].sum()
   Imports_Idn['Total_Idn_Tarjihi_year']=Imports_Idn[Imports_Idn['Tarjihi']==1]['Dollars_Tot'].sum()


   Exports_Idn = Exports_Idn.rename(columns={c: c+f'_{year}' for c in Exports_Idn.columns if c not in ['Tcode', 'Tdesc','Tarjihi']})
   Imports_Idn = Imports_Idn.rename(columns={c: c+f'_{year}' for c in Imports_Idn.columns if c not in ['Tcode', 'Tdesc','Tarjihi']})


   if year==baseyear:
      Exports_Idn_Series = Exports_Idn
      Imports_Idn_Series = Imports_Idn
   else:
      Exports_Idn_Series = Exports_Idn_Series.merge(Exports_Idn,on=['Tcode','Tdesc','Tarjihi'],how='outer')
      Imports_Idn_Series = Imports_Idn_Series.merge(Imports_Idn,on=['Tcode','Tdesc','Tarjihi'],how='outer')


   
```
