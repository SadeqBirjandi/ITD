```python
import pandas as pd
import numpy as np
import os
```

```python
os.chdir("C:\TradeGroup")
my_wd = os.getcwd()
new_directory = my_wd
```

```python
year = 1380
for year in range(1380,1403):

    Country_Data = pd.read_excel(f'{my_wd}/Data/Country_Data/Y{year}CountryTrade.xlsx')

    Country_Data_New = Country_Data[['Country','Trade_Volume_Share']]

    Country_Data_New.sort_values(by=['Trade_Volume_Share'],ascending=False, inplace=True)

    Country_Data_Columns = [f'Country{year}',f'Trade_Share{year}']

    Rank = range(1,11)


    Country_Data_New.columns = Country_Data_Columns

    Country_Data_New_10 = Country_Data_New[:10]

    Country_Data_New_10['Rank']= Rank

    if year == 1380:
        Countries = Country_Data_New_10
    else:
        Countries = Countries.merge(Country_Data_New_10, on=['Rank'] , how='outer')


Countries        



```

```python
Countries.to_excel(f'C:\TradeGroup\Reports\TenFirstCountries.xlsx',index=False)
```
