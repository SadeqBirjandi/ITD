```python
import pandas as pd
import numpy as np
import os
```

```python
os.chdir("C:\TradeGroup")
my_wd = os.getcwd()
new_directory = my_wd

def create_folder(folder_path):
    if not os.path.exists(f'{new_directory}/{folder_path}'):
        os.makedirs(f'{new_directory}/{folder_path}')
    else:
        print(f'Already Exists')

create_folder('Data/Indices')
```

```python
baseyear = 1392
for year in range(1392,1403):
    Data = pd.read_excel(f"{my_wd}/Data/Cleaned_Data/Y{year}Exports.xlsx")
    Data = Data.dropna(subset=['Tcode'])
    colname = f'Dollars_Exp_{year}'
    Data['HS6'] = Data['Tcode']//100
    Data_HS6 = Data.groupby('HS6').agg(
    Dollars = (f'Dollars','sum')
    ).reset_index()



    #Data_HS6.columns.str.replace("Dollars",f'Dollars_Exp_{year}',inplace=True)
    Data_HS6.columns = Data_HS6.columns.str.replace("Dollars", f"Dollars_Exp_{year}")
    Data_HS6[f'Sign_{year}']=np.where(Data_HS6[f"Dollars_Exp_{year}"]>10000,1,0)
    Data_HS6.to_excel(f'{my_wd}/Data/Indices/Exports_{year}_HS6.xlsx', index = False)    
    if year == baseyear:
        TimeSeries = Data_HS6[Data_HS6[f'Sign_{year}']==1]
        TimeSeries['Persistency']=TimeSeries[f'Sign_{year}']
        #TimeSeries = TimeSeries[TimeSeries['Persistency']==1]
        counter= 0
    else:
        counter = counter + 1
        TimeSeries = TimeSeries.merge(Data_HS6,on="HS6",how='left')
        TimeSeries.fillna(0,inplace=True) 
        TimeSeries['Persistency_Helper']= TimeSeries['Persistency']
        TimeSeries['Persistency']=np.where((TimeSeries['Persistency_Helper']== counter) &
                                            (TimeSeries[f'Sign_{year}']>0),
                                             TimeSeries['Persistency_Helper']+1,
                                             TimeSeries['Persistency_Helper']) 
```

```python
TimeSeries.to_excel(f'{my_wd}/Data/Indices/Persistency_HS6_Detailed.xlsx',index = False)


TimeSeries_Uniques = TimeSeries.groupby('Persistency').agg(
    Persistency_Count = ('Persistency','count')
).reset_index()

TimeSeries_Uniques['Countinues_Years'] = np.where(TimeSeries_Uniques['Persistency']>0,TimeSeries_Uniques['Persistency']-1,0)

TimeSeries_Uniques.to_excel(f'{my_wd}/Data/Indices/Persistency_HS6.xlsx',index = False)

TimeSeries_Uniques
```
