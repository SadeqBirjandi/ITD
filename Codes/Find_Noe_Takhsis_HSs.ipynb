```python
import pandas as pd
import numpy as np
import os

os.chdir("C:\TradeGroup")
my_wd = os.getcwd()
new_directory = my_wd
```

```python
Tamin_Arz = pd.read_excel(f'C:/TradeGroup/Reports/ArzTakhsisi_Alpha.xlsx')

Tamin_Arz
```

```python
Tamin_Arz_By_Group = Tamin_Arz.groupby(['Tcode',' Takhsis_Type']).agg(
    MDollars =('MDollars','sum')
).reset_index()

Tamin_Arz_By_Group['Share'] = Tamin_Arz_By_Group['MDollars']/Tamin_Arz_By_Group.groupby('Tcode')['MDollars'].transform('sum')

Tamin_Arz_By_Group
#
# Tamin_Arz_By_Group.to_excel(f'C:/TradeGroup/Reports/Codes_Repeated_Varedat_Arz_Type_Shares.xlsx')
```

```python
max_idx = Tamin_Arz_By_Group.groupby('Tcode')['Share'].idxmax()
max_idx = max_idx.dropna()  # Remove NaN values from max_idx

Tamin_Arz_By_Group.reset_index(inplace=True, drop=True)
filtered_df_new = Tamin_Arz_By_Group.loc[max_idx]
filtered_df_new
```

```python
Tamin_Arz_By_Group['IsMax'] = Tamin_Arz_By_Group.groupby('Tcode')['Share'].transform(lambda x: x.eq(x.max()).astype(int))
Tamin_Arz_By_Group_unique = Tamin_Arz_By_Group[Tamin_Arz_By_Group['IsMax']==1]

Tamin_Arz_By_Group_unique_rows=Tamin_Arz_By_Group_unique[['Tcode',' Takhsis_Type']]

Tamin_Arz_By_Group_unique_rows.to_excel(f'C:/TradeGroup/Data/Dashboard/Tkhsis_Type_for_Dashboard.xlsx', index=False)
```
