```python
import pandas as pd
import numpy as np
import os


os.chdir("C:\TradeGroup")
my_wd = os.getcwd()
new_directory = my_wd
```

```python
HS4_Imports_TimeSeries = pd.read_excel(f"{my_wd}/Data/HS_Data/HS4_Imports_TimeSeries.xlsx")

HS4_Exports_TimeSeries = pd.read_excel(f"{my_wd}/Data/HS_Data/HS4_Exports_TimeSeries.xlsx")


Red_Meat_Codes = pd.read_excel(f"{my_wd}/Reports/RedMeat.xlsx")

HS4_RedMeat_Imports = HS4_Imports_TimeSeries[HS4_Imports_TimeSeries['HS4'].isin(Red_Meat_Codes['HS4'])]

HS4_RedMeat_Exports = HS4_Exports_TimeSeries[HS4_Exports_TimeSeries['HS4'].isin(Red_Meat_Codes['HS4'])]

new_column_names_Imports = ['Imp_' + col if i != 0 else col for i, col in enumerate(HS4_RedMeat_Imports.columns)]
HS4_RedMeat_Imports.columns = new_column_names_Imports

new_column_names_Exports = ['Exp_' + col if i != 0 else col for i, col in enumerate(HS4_RedMeat_Exports.columns)]
HS4_RedMeat_Exports.columns = new_column_names_Exports




HS4_RedMeat=HS4_RedMeat_Imports.merge(HS4_RedMeat_Exports,on="HS4",how='outer')
Red_Meat_Trade=pd.DataFrame({"HS4" : HS4_RedMeat['HS4']})
for year in range(1390,1403):
    Red_Meat_Trade[f'Balance_{year}']=HS4_RedMeat[f'Exp_{year}']-HS4_RedMeat[f'Imp_{year}']
 

Red_Meat_Trade
```

```python
Iran_Sheep_Exports = pd.read_excel(f"{my_wd}/Reports/Iran_Sheep_Exports_Yearly.xlsx")
Iran_Bovine_Exports = pd.read_excel(f"{my_wd}/Reports/Iran_Bovine_Exports_Yearly.xlsx")

def convert_amount(amount_str):
    if 'M' in amount_str:
        return float(amount_str.replace('$', '').replace('M', '')) * 1e6
    elif 'k' in amount_str:
        return float(amount_str.replace('$', '').replace('k', '')) * 1e3
    else:
        return float(amount_str.replace('$', ''))
    
Iran_Sheep_Exports['Exports']= Iran_Sheep_Exports['Current Gross Export'].apply(convert_amount)
Iran_Bovine_Exports['Exports']= Iran_Bovine_Exports['Current Gross Export'].apply(convert_amount)

Iran_Sheep_Exports_wide = Iran_Sheep_Exports.pivot(index='Name', columns='Year', values='Exports')
Iran_Sheep_Exports_wide.fillna(0,inplace=True)
Iran_Bovine_Exports_wide = Iran_Bovine_Exports.pivot(index='Name', columns='Year', values='Exports')
Iran_Bovine_Exports_wide.fillna(0,inplace=True)

Exports= pd.DataFrame({'Sheep': Iran_Sheep_Exports_wide.sum(axis=0),'Bovine': Iran_Bovine_Exports_wide.sum(axis=0)})
Exports.fillna(0,inplace=True)

Exports['Year']=range(1995,2022)
Exports


Exports = Exports.reset_index(drop=True)
Exports
```

```python
Red_Meat_Exports = pd.read_excel(f"{my_wd}/Reports/RedMeat_HS4.xlsx",sheet_name="Sheet3")
Red_Meat_Exports.rename(columns={"Bovine":"Bovine_Gomrok","Sheep":"Sheep_Gomrok"},inplace=True)
Red_Meat_Exports
```

```python
#Exports.loc[Exports[Year]==2022,"Bovine"]=0
Harvard_Shamsi = pd.DataFrame({"Year":range(1390,1402),"Bovine_Harvard":"NaN","Sheep_Harvard":"NaN"})
for year in range(2011,2021):
    Harvard_Shamsi.loc[Harvard_Shamsi["Year"]==(year-621),'Bovine_Harvard']=3*float(Exports.loc[Exports["Year"]==year,'Bovine'])/4 + float(Exports.loc[Exports["Year"]==(year+1),'Bovine'])/4
    Harvard_Shamsi.loc[Harvard_Shamsi["Year"]==(year-621),'Sheep_Harvard']=3*float(Exports.loc[Exports["Year"]==year,'Sheep'])/4 + float(Exports.loc[Exports["Year"]==(year+1),'Sheep'])/4

Harvard_Shamsi
```

```python
Merged_Exports=Red_Meat_Exports.merge(Harvard_Shamsi,on="Year",how="right")
#.dropna(inplace=True)
Merged_Exports=Merged_Exports[Merged_Exports["Sheep_Harvard"]!="NaN"]
Merged_Exports["Sheep_Harvard_More"]=Merged_Exports['Sheep_Harvard']-Merged_Exports['Sheep_Gomrok']
Merged_Exports["Bovine_Harvard_More"]=Merged_Exports['Bovine_Harvard']-Merged_Exports['Bovine_Gomrok']

Merged_Exports.to_excel(f"{my_wd}/Reports/Exports_Harvard_Gomrok.xlsx",index=False)
Merged_Exports
```

```python
Oman_Sheep_Imports = pd.read_excel(f"{my_wd}/Reports/Others_Sheep_Imports.xlsx")

def convert_amount(amount_str):
    if 'M' in amount_str:
        return float(amount_str.replace('$', '').replace('M', '')) * 1e6
    elif 'k' in amount_str:
        return float(amount_str.replace('$', '').replace('k', '')) * 1e3
    else:
        return float(amount_str.replace('$', ''))
    
Oman_Sheep_Imports['Imports']= Oman_Sheep_Imports['Current Gross Import'].apply(convert_amount)

Iran_Sheep_Exports_wide = Iran_Sheep_Exports.pivot(index='Name', columns='Year', values='Exports')
Iran_Sheep_Exports_wide.fillna(0,inplace=True)
Iran_Bovine_Exports_wide = Iran_Bovine_Exports.pivot(index='Name', columns='Year', values='Exports')
Iran_Bovine_Exports_wide.fillna(0,inplace=True)

Exports= pd.DataFrame({'Sheep': Iran_Sheep_Exports_wide.sum(axis=0),'Bovine': Iran_Bovine_Exports_wide.sum(axis=0)})
Exports.fillna(0,inplace=True)

Exports['Year']=range(1995,2022)
Exports


Exports = Exports.reset_index(drop=True)
Exports
```

```python
Oman_Sheep_Imports = pd.read_excel(f"{my_wd}/Reports/Others_RedMeat_Imports.xlsx",sheet_name="Oman")
UAE_Sheep_Imports = pd.read_excel(f"{my_wd}/Reports/Others_RedMeat_Imports.xlsx",sheet_name="UAE")
Iraq_Sheep_Imports = pd.read_excel(f"{my_wd}/Reports/Others_RedMeat_Imports.xlsx",sheet_name="Iraq")
Qatar_Sheep_Imports = pd.read_excel(f"{my_wd}/Reports/Others_RedMeat_Imports.xlsx",sheet_name="Qatar")
Kuwait_Sheep_Imports = pd.read_excel(f"{my_wd}/Reports/Others_RedMeat_Imports.xlsx",sheet_name="Kuwait")




dataframes = {"Oman": Oman_Sheep_Imports,"UAE": UAE_Sheep_Imports,
              #"Iraq":Iraq_Sheep_Imports,
                "Qatar": Qatar_Sheep_Imports,
                "Kuwait": Kuwait_Sheep_Imports}

def convert_amount(amount_str):
    if 'M' in amount_str:
        return float(amount_str.replace('$', '').replace('M', '')) * 1e6
    elif 'k' in amount_str:
        return float(amount_str.replace('$', '').replace('k', '')) * 1e3
    else:
        return float(amount_str.replace('$', ''))
    
Panel_Countries = pd.DataFrame({"Year":range(1995,2022)})    
for Country, df_name in dataframes.items():
    #df_name=Country+"_Sheep_Imports"
    #df_name =pd.read_excel(f"{my_wd}/Reports/Others_RedMeat_Imports.xlsx",sheet_name=Country)

    #df_name=Country+"_Sheep_Imports"
    df_name['Imports']= df_name['Current Gross Import'].apply(convert_amount)    
    
    df_name['Imports']= df_name['Current Gross Import'].apply(convert_amount)

    #df_name_wide = df_name+"_wide"
    df_name_wide = df_name.pivot(index='Name', columns='Year', values='Imports')
    df_name_wide.fillna(0,inplace=True)
    df_name_wide.reset_index(inplace=True)

    #df_name_Iran_Share = df_name+"_Iran_Share"
    df_name_Iran_Share=pd.DataFrame(data={'Year': range(1995,2022),f'Imports_{Country}':0,f'Iran_Imports_{Country}':0,f'IranShare_{Country}':0})
    for year in range(1995,2022):
        df_name_Iran_Share.loc[df_name_Iran_Share['Year']==year,f'Imports_{Country}'] = df_name_wide[year].sum()
        df_name_Iran_Share.loc[df_name_Iran_Share['Year']==year,f'Iran_Imports_{Country}'] = df_name_wide.loc[df_name_wide['Name']=="Iran",year].values
    df_name_Iran_Share[f'IranShare_{Country}']=np.where(df_name_Iran_Share[f'Iran_Imports_{Country}']!=0,df_name_Iran_Share[f'Iran_Imports_{Country}']/df_name_Iran_Share[f'Imports_{Country}'],0)

    Panel_Countries= Panel_Countries.merge(df_name_Iran_Share,on="Year")
        

Panel_Countries.to_excel(f"{my_wd}/Reports/Dam_Imports_From_Iran.xlsx",index=False)
```
