```python
year = 1390
for year in range(1390,1403):
    exports_file_path = f"{my_wd}/Data/Cleaned_Data/Y{year}Exports.xlsx"
    imports_file_path = f"{my_wd}/Data/Cleaned_Data/Y{year}Imports.xlsx"
    Exports = pd.read_excel(exports_file_path)
    Imports = pd.read_excel(imports_file_path)

    Exports.dropna(axis=0,inplace=True)
    Imports.dropna(axis=0,inplace=True)

    Exports['Tcode'] = Exports['Tcode'].astype('int')
    Imports['Tcode'] = Imports['Tcode'].astype('int')




    Exports['HS4'] = Exports['Tcode'].astype('str').str.slice(0,-4)
    Exports['HS6'] = Exports['Tcode'].astype('str').str.slice(0,-2)

    Exports_Gav=Exports[Exports['HS4']=="201"]['Weight'].sum()
    Exports_Gav_Yakh=Exports[Exports['HS4']=="202"]['Weight'].sum()
    Exports_Goos = Exports[(Exports['HS4']=="204") & (Exports['Tcode']!=2045000)]['Weight'].sum()
    Exports_Boz = Exports[(Exports['Tcode']==2045000)]['Weight'].sum()
    Exports_ahsham=Exports[Exports['HS4']=="206"]['Weight'].sum()


    Exports_Gav_Zendeh = Exports[Exports['HS4']=="102"]['Weight'].sum()
    Exports_Goos_Zendeh = Exports[(Exports['HS4']=="104")&(Exports['HS6']!="10420")]['Weight'].sum()
    Exports_Boz_Zendeh = Exports[Exports['HS6']=="10420"]['Weight'].sum()



    Imports['HS4'] = Imports['Tcode'].astype('str').str.slice(0,-4)
    Imports['HS6'] = Imports['Tcode'].astype('str').str.slice(0,-2)

    Imports_Gav=Imports[Imports['HS4']=="201"]['Weight'].sum()
    Imports_Gav_Yakh=Imports[Imports['HS4']=="202"]['Weight'].sum()
    Imports_Goos = Imports[(Imports['HS4']=="204") & (Imports['Tcode']!=2045000)]['Weight'].sum()
    Imports_Boz = Imports[(Imports['Tcode']==2045000)]['Weight'].sum()
    Imports_ahsham=Imports[Imports['HS4']=="206"]['Weight'].sum()

    Imports_Gav_Zendeh = Imports[Imports['HS4']=="102"]['Weight'].sum()
    Imports_Goos_Zendeh = Imports[(Imports['HS4']=="104")&(Imports['HS6']!="10420")]['Weight'].sum()
    Imports_Boz_Zendeh = Imports[Imports['HS6']=="10420"]['Weight'].sum()

    Year = year

    data = {'Year': year,
            'Exports_Gav_Zendeh': Exports_Gav_Zendeh,
            'Exports_Goos_Zendeh': Exports_Goos_Zendeh,
            'Exports_Boz_Zendeh': Exports_Boz_Zendeh,
            
            'Imports_Gav_Zendeh':Imports_Gav_Zendeh,
            'Imports_Goos_Zendeh':Imports_Goos_Zendeh,
            'Imports_Boz_Zendeh': Imports_Boz_Zendeh,
            
            'Exports_Gav':Exports_Gav,
            'Exports_Gav_Yakh':Exports_Gav_Yakh,
            'Exports_Goos':Exports_Goos,
            'Exports_Boz':Exports_Boz,
            'Exports_ahsham':Exports_ahsham,

            'Imports_Gav':Imports_Gav,
            'Imports_Gav_Yakh':Imports_Gav_Yakh,
            'Imports_Goos':Imports_Goos,
            'Imports_Boz':Imports_Boz,
            'Imports_ahsham':Imports_ahsham
            }
    Trade = pd.DataFrame(data,index=[str(year)])
    if year == 1390:
        Trade_Series = Trade
    else:
        Trade_Series = pd.concat([Trade_Series,Trade])    

```
