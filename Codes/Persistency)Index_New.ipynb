```python
import pandas as pd
import numpy as np
import os
```

```python
os.chdir("C:\TradeGroup")
my_wd = os.getcwd()
new_directory = my_wd

def create_folder(folder_path):
    if not os.path.exists(f'{new_directory}/{folder_path}'):
        os.makedirs(f'{new_directory}/{folder_path}')
    else:
        print(f'Already Exists')

create_folder('Data/Indices')
```

```python
baseyear = 1392
for year in range(1392,1403):
    
    Data = pd.read_excel(f"{my_wd}/Data/Cleaned_Data/Y{year}Exports.xlsx")
    Data = Data.dropna(subset=['Tcode'])
    colname = f'Dollars_Exp_{year}'
    Data_HS8 = Data.groupby('Tcode').agg(
    Dollars = (f'Dollars','sum')
    ).reset_index()

    Data_HS8.columns = Data_HS8.columns.str.replace("Dollars", f"Dollars_Exp_{year}")
    Data_HS8[f'Sign_{year}']=np.where(Data_HS8[f"Dollars_Exp_{year}"]>10000,1,0)
    


    if year==baseyear:

        Data_HS8['Years_in']=1
        Data_HS8_Series=Data_HS8
        Data_HS8_Series['Terminate_Year']=np.nan
    else:
        Data_HS8_Series.loc[Data_HS8_Series['Years_in'].isna()]=1
        Data_HS8_Series = Data_HS8_Series.merge(Data_HS8,on="Tcode",how='outer')
        Data_HS8_Series.loc[Data_HS8_Series[f'Sign_{year}'].isna(), f'Sign_{year}'] = 0
        Data_HS8_Series.loc[Data_HS8_Series[f'Sign_{year-1}'].isna(), f'Sign_{year-1}'] = 0
        Data_HS8_Series.loc[
            (Data_HS8_Series[f'Sign_{year-1}'] == 1) & 
            (Data_HS8_Series[f'Sign_{year}'] == 0) & 
            Data_HS8_Series['Terminate_Year'].isna(),
            'Terminate_Year'
            ] = year - 1
        Data_HS8_Series.loc[(Data_HS8_Series[f'Years_in'].isna()),'Years_in']=1
        
        Data_HS8_Series.loc[
             (Data_HS8_Series[f'Sign_{year-1}'] == 1) & 
             (Data_HS8_Series[f'Sign_{year}'] == 1) & 
             Data_HS8_Series['Terminate_Year'].isna(),
            'Years_in'
            ] += 1
        #Data_HS8_Series.loc[(Data_HS8_Series[f'Sign_{year}']==1) & (Data_HS8_Series['Terminate_Year'].isna() ),'Years_in']=1



Data_HS8_Series   

          
```

```python
Data_HS8_Series.to_excel(f'{my_wd}/Data/Indices/Persistency_News_Detailed.xlsx',index = False)
```
