```python
import pandas as pd
import numpy as np
import os
```

```python
# Set the working directory
os.chdir("C:\TradeGroup")
my_wd = os.getcwd()
new_directory = my_wd

def create_folder(folder_path):
    if not os.path.exists(f'{new_directory}/{folder_path}'):
        os.makedirs(f'{new_directory}/{folder_path}')
    else:
        print(f'Already Exists')

create_folder('Data/Indices')
```

```python
Economic_Openness_Path = os.path.join(my_wd"Data/Indices/Economic_Openness.xlsx")


Economic_Openness = pd.read_excel(Economic_Openness_Path)

Economic_Openness
```

```python
ranked_df = Economic_Openness.set_index("Country Name").rank(axis=0, method='average', numeric_only=True, na_option="keep",
                                                             ascending=False).reset_index()

ranked_df_Iran=ranked_df[ranked_df['Country Name']=="Iran, Islamic Rep."]


Iran_Economic_Openness= Economic_Openness[Economic_Openness['Country Name']=="Iran, Islamic Rep."]

CountriesNumbers=Iran_Economic_Openness.copy()

CountriesNumbers_columns=CountriesNumbers.columns[1:]

for i in CountriesNumbers_columns:
    CountriesNumbers[i]=Economic_Openness[i].count()

CountriesNumbers['Country Name']= "Total Observations"

CountriesNumbers
```

```python
EO=pd.concat([Iran_Economic_Openness,ranked_df_Iran])

EO=pd.concat([EO,CountriesNumbers])

EO.to_excel(f'{my_wd}/Data/Indices/Economic_Openness_Modified.xlsx', index=False)
```
