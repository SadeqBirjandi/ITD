```python
import pandas as pd
import numpy as np
import os
```

```python
year = 1401
baseyear=1389

os.chdir("C:\TradeGroup")
my_wd = os.getcwd()
new_directory = my_wd
Wanted_Codes=pd.read_excel(f'{my_wd}\MrFamilMohammadi.xlsx')

for year in range(1389,1403):
    print(year)
    Exports_Iran=pd.read_excel(f'{my_wd}\Data\Cleaned_Data\Y{year}Exports.xlsx')
    Exports_Iran=Exports_Iran[Exports_Iran['Tcode'].notnull()]


    Exports_Iran['HS6']=Exports_Iran['Tcode']//100

    Exports_Iran_HS6=Exports_Iran.groupby("HS6").agg(
        TDesc=("Tdesc","first"),
        Weight=("Weight","sum"),
        Rials=("Rials","sum"),
        Dollars=("Dollars","sum")
        ).reset_index()
    Exports_Iran_HS6['HS4_Final']=Exports_Iran_HS6['HS6']//100
    Exports_Iran_HS6
    Exports_Iran_HS6.loc[Exports_Iran_HS6["HS6"]==80610,"HS4_Final"]=8061
    Exports_Iran_HS6.loc[Exports_Iran_HS6["HS6"]==80620,"HS4_Final"]=8062

    Exports_Limited=Wanted_Codes.merge(Exports_Iran_HS6,on="HS4_Final",how="left")


    Exports_Limited['HS4_Total'] = Exports_Limited.groupby('HS4_Final')['Dollars'].transform('sum')
    Exports_Limited['HS6_Total'] = Exports_Limited.groupby('HS6')['Dollars'].transform('sum')
    Exports_Limited[f'Share_{year}'] = Exports_Limited['HS6_Total'] / Exports_Limited['HS4_Total']
    Exports_Limited=Exports_Limited.dropna(subset=['Weight'])


    idx_max_share = Exports_Limited.groupby('HS4_Final')[f'Share_{year}'].idxmax()
    Exports_Limited_max_share = Exports_Limited.loc[idx_max_share]
    Exports_Limited_max_share=Exports_Limited_max_share.sort_index()
    Exports_Limited_max_share[f'Price_{year}']=Exports_Limited_max_share['Dollars']/Exports_Limited_max_share['Weight']

    Exports_With_Prices=Exports_Limited_max_share[['Group','Comm','HS6','HS4_Final',f'Price_{year}']]
    Exports_With_Shares=Exports_Limited_max_share[['Group','Comm','HS6','HS4_Final',f'Share_{year}']]



    if year==baseyear:
        Exports_With_Prices_Series=Exports_With_Prices
        Exports_With_Shares_Series=Exports_With_Shares

    else:
        Exports_With_Prices_Series=Exports_With_Prices_Series.merge(Exports_With_Prices,on=['Group','Comm','HS6','HS4_Final'],how='outer')
        Exports_With_Shares_Series=Exports_With_Shares_Series.merge(Exports_With_Shares,on=['Group','Comm','HS6','HS4_Final'],how='outer')
    


    Exports_Limited_max_share.to_excel(f'{my_wd}\Wanted_Data\\Com61_{year}Export.xlsx')   
    print(f'{year} Completed')


Exports_With_Prices_Series.to_excel(f'{my_wd}\Wanted_Data\Com61_Series.xlsx')
Exports_With_Shares_Series.to_excel(f'{my_wd}\Wanted_Data\Com61_Series_Share.xlsx')





```
