```python
# Clear all variables in the workspace
locals().clear()
import os
import pandas as pd

# Set the working directory
os.chdir("C:\TradeGroup")
my_wd = os.getcwd()
year = 1401


# Read Imports data
Imports = pd.read_excel(os.path.join(my_wd, "Imports_ModelingGroup.xlsx"), sheet_name="weight")
Imports_Rials = pd.read_excel(os.path.join(my_wd, "Imports_ModelingGroup.xlsx"), sheet_name="Rial_Price")
Imports_Dollars = pd.read_excel(os.path.join(my_wd, "Imports_ModelingGroup.xlsx"), sheet_name="Dollar_Price")

# Read Exports data
Exports = pd.read_excel(os.path.join(my_wd, "Exports_ModelingGroup.xlsx"), sheet_name="weight")
Exports_Rials = pd.read_excel(os.path.join(my_wd, "Exports_ModelingGroup.xlsx"), sheet_name="Rial_Price")
Exports_Dollars = pd.read_excel(os.path.join(my_wd, "Exports_ModelingGroup.xlsx"), sheet_name="Dollar_Price")
```

```python
def create_folder(folder_path):
    if not os.path.exists(f'{new_directory}/{folder_path}'):
        os.makedirs(f'{new_directory}/{folder_path}')
    else:
        print(f'Already Exists')
```

```python
new_directory = my_wd
create_folder('Data/Monthly_Data')

# Extract selected columns and replace NaN with 0 for Imports data
Imports_mod = Imports.iloc[:, [1, 2, 3] + list(range(23, 202))]
Imports_mod = Imports_mod.fillna(0)

# Extract selected columns and replace NaN with 0 for Imports_Rials data
Imports_Rials_mod = Imports_Rials.iloc[:, [1, 2, 3] + list(range(23, 202))]
Imports_Rials_mod = Imports_Rials_mod.fillna(0)

# Extract selected columns and replace NaN with 0 for Imports_Dollars data
Imports_Dollars_mod = Imports_Dollars.iloc[:, [1, 2, 3] + list(range(23, 202))]
Imports_Dollars_mod = Imports_Dollars_mod.fillna(0)

# Extract selected columns and replace NaN with 0 for Exports data
Exports_mod = Exports.iloc[:, [1, 2, 3] + list(range(23, 202))]
Exports_mod = Exports_mod.fillna(0)

# Extract selected columns and replace NaN with 0 for Exports_Rials data
Exports_Rials_mod = Exports_Rials.iloc[:, [1, 2, 3] + list(range(23, 202))]
Exports_Rials_mod = Exports_Rials_mod.fillna(0)

# Extract selected columns and replace NaN with 0 for Exports_Dollars data
Exports_Dollars_mod = Exports_Dollars.iloc[:, [1, 2, 3] + list(range(23, 202))]
Exports_Dollars_mod = Exports_Dollars_mod.fillna(0)

# Get column names
My_colnames = Imports_mod.columns.tolist()
```

```python
for colmoi in My_colnames[3:182]:
    # Process Imports data
    IMW_col = Imports_mod[['Tariff_description', 'HS6Desc', 'HS_8d', colmoi]].copy()
    IMW_col.columns = ['Tariff_description', 'HS6Desc', 'HS_8d', f'{colmoi}_Weight_Imp']

    IMR_col = Imports_Rials_mod[['Tariff_description', 'HS6Desc', 'HS_8d', colmoi]].copy()
    IMR_col.columns = ['Tariff_description', 'HS6Desc', 'HS_8d', f'{colmoi}_Rials_Imp']

    IMD_col = Imports_Dollars_mod[['Tariff_description', 'HS6Desc', 'HS_8d', colmoi]].copy()
    IMD_col.columns = ['Tariff_description', 'HS6Desc', 'HS_8d', f'{colmoi}_Dollars_Imp']

    IM_col = pd.merge(IMW_col, IMR_col, on=['Tariff_description', 'HS6Desc', 'HS_8d'], how='outer')
    IM_col = pd.merge(IM_col, IMD_col, on=['Tariff_description', 'HS6Desc', 'HS_8d'], how='outer')

    IM_col['Tariff_description'].replace(0, pd.NA, inplace=True)
    IM_col['HS6Desc'].replace(0, pd.NA, inplace=True)

    # Process Exports data
    EMW_col = Exports_mod[['Tariff_description', 'HS6Desc', 'HS_8d', colmoi]].copy()
    EMW_col.columns = ['Tariff_description', 'HS6Desc', 'HS_8d', f'{colmoi}_Weight_Exp']

    EMR_col = Exports_Rials_mod[['Tariff_description', 'HS6Desc', 'HS_8d', colmoi]].copy()
    EMR_col.columns = ['Tariff_description', 'HS6Desc', 'HS_8d', f'{colmoi}_Rials_Exp']

    EMD_col = Exports_Dollars_mod[['Tariff_description', 'HS6Desc', 'HS_8d', colmoi]].copy()
    EMD_col.columns = ['Tariff_description', 'HS6Desc', 'HS_8d', f'{colmoi}_Dollars_Exp']

    EM_col = pd.merge(EMW_col, EMR_col, on=['Tariff_description', 'HS6Desc', 'HS_8d'], how='outer')
    EM_col = pd.merge(EM_col, EMD_col, on=['Tariff_description', 'HS6Desc', 'HS_8d'], how='outer')

    EM_col['Tariff_description'].replace(0, pd.NA, inplace=True)
    EM_col['HS6Desc'].replace(0, pd.NA, inplace=True)

    # Merge Imports and Exports data
    M_col = pd.merge(EM_col, IM_col, on=['Tariff_description', 'HS6Desc', 'HS_8d'], how='outer')
    last_6_columns = M_col.columns[-6:]
    M_col[last_6_columns] = M_col[last_6_columns].fillna(0)
    # Save to CSV
    M_col.to_excel(os.path.join(my_wd, 'Data', 'Monthly_Data', f'M{colmoi}.xlsx'), index=False)
```
