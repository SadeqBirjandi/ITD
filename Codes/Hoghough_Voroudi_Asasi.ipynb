```python
import pandas as pd
import numpy as np
import os
import openpyxl
import xlrd
```

```python
# Set the working directory
os.chdir("C:\TradeGroup")
my_wd = os.getcwd()
new_directory = my_wd

def create_folder(folder_path):
    if not os.path.exists(f'{new_directory}/{folder_path}'):
        os.makedirs(f'{new_directory}/{folder_path}')
    else:
        print(f'Already Exists')

create_folder('Data/Indices')
```

```python
Gorouh_Kalayi = pd.read_excel(f'C:/TradeGroup/Data/Indices/Kala_Asasi.xlsx')

Nerkh_Tarefe = pd.read_excel(f'C:/TradeGroup/Reports/3 ماه اول/Nerkh_Tarefe_Part2.xlsx',sheet_name="1403_03")


Nerkh_Tarefe =Nerkh_Tarefe[Nerkh_Tarefe['کد تعرفه']!='جمع کل']
Nerkh_Tarefe['کد تعرفه']= Nerkh_Tarefe['کد تعرفه'].astype('float')


new_column_names = ['Tcode', 'Weight', 'Value_R','Value_D','HV_041_R','Other_042_R',
                    'NN_043_R','TA_044_R','TM_045_R','J_046_R','M_047_R','A_048_R',
                    'THV_049_R','HSum_R','HPerc_R']

Nerkh_Tarefe = Nerkh_Tarefe.rename(columns=dict(zip(Nerkh_Tarefe.columns, new_column_names)))

Nerkh_Tarefe
```

```python
my_data = Nerkh_Tarefe.merge(Gorouh_Kalayi,on='Tcode',how='left')

my_data['HS6'] = my_data['Tcode']//100

HS6s = pd.read_excel(f'C:/TradeGroup/Data/Dashboard/HSs_complete_Final31.xlsx')

my_data_new = my_data.merge(HS6s,on='HS6',how='left')

my_data_new.to_excel(f'C:/TradeGroup\Reports/Hoghough_Voroudi_HS8.xlsx',index=False)

```

```python
Hoghough_Voroudi_Noe_Olaviat = my_data_new.groupby('Priority').agg(
    Import_D = ('Value_D','sum'),
    Import_R = ('Value_R','sum'),
    Hoghough_Voroudi_R = ('HSum_R','sum')
   # Hoghough_Voroudi_Perc = ('HPerc_R','mean')
).reset_index()


Hoghough_Voroudi_Noe_Olaviat['Darsad_Varedat']=round(100*Hoghough_Voroudi_Noe_Olaviat['Import_R']/Hoghough_Voroudi_Noe_Olaviat['Import_R'].sum(),3)
Hoghough_Voroudi_Noe_Olaviat['H_Perc'] = round(100*Hoghough_Voroudi_Noe_Olaviat['Hoghough_Voroudi_R']/Hoghough_Voroudi_Noe_Olaviat['Import_R'],3)


Hoghough_Voroudi_Noe_Olaviat.to_excel(f'C:/TradeGroup/Reports/Hoghough_Voroudi_Priority.xlsx',index=False)
```

```python
my_data_new['isPrior']=np.where(my_data_new['Priority']=="21",1,0)

Hoghough_Voroudi_Prior = my_data_new.groupby('isPrior').agg(
    Import_D = ('Value_D','sum'),
    Import_R = ('Value_R','sum'),
    Hoghough_Voroudi_R = ('HSum_R','sum')
   # Hoghough_Voroudi_Perc = ('HPerc_R','mean')
).reset_index()
Hoghough_Voroudi_Prior['Darsad_Varedat']=round(100*Hoghough_Voroudi_Prior['Import_R']/Hoghough_Voroudi_Prior['Import_R'].sum(),3)
Hoghough_Voroudi_Prior['H_Perc'] =  round(100*Hoghough_Voroudi_Prior['Hoghough_Voroudi_R']/Hoghough_Voroudi_Prior['Import_R'],3)

Hoghough_Voroudi_Prior.to_excel(f'C:/TradeGroup/Reports/Hoghough_Voroudi_isPrior.xlsx',index=False)


```

```python
my_data_new_isnotprior = my_data_new[my_data_new['isPrior']==0]

my_data_new_isnotprior

HS6s = my_data_new_isnotprior.groupby('HS6').agg(
    HS2_F = ('HS2_F','first'),
    HS4_F = ('HS4_F','first'),
    HS6_F = ('HS6_F','first'),
    Import_R = ('Value_R','sum'),
    Import_D = ('Value_D','sum'),

    Hoghough_Voroudi_R = ('HSum_R','sum')
).reset_index()
HS6s['Darsad_Varedat']=round(100*HS6s['Import_R']/HS6s['Import_R'].sum(),3)
HS6s['H_Perc'] = round(100*HS6s['Hoghough_Voroudi_R']/HS6s['Import_R'],3)

HS6s.to_excel(f'C:/TradeGroup/Reports/Hoghough_Voroudi_isnotPrior_HS6.xlsx',index=False)






```

```python
my_data_new_isnotprior['HS4']=my_data_new_isnotprior['Tcode']//10000

my_data_new_isnotprior['HS2']=my_data_new_isnotprior['Tcode']//1000000

HS4s = my_data_new_isnotprior.groupby('HS4').agg(
    HS4_F = ('HS4_F','first'),
    HS6_F = ('HS6_F','first'),
    Import_R = ('Value_R','sum'),
    Import_D = ('Value_D','sum'),

    Hoghough_Voroudi_R = ('HSum_R','sum')
).reset_index()

HS4s['Darsad_Varedat']=round(100*HS4s['Import_R']/HS4s['Import_R'].sum(),3)
HS4s['H_Perc'] = round(100*HS4s['Hoghough_Voroudi_R']/HS4s['Import_R'],3)

HS4s.to_excel(f'C:/TradeGroup/Reports/Hoghough_Voroudi_isnotPrior_HS4.xlsx',index=False)




HS2s = my_data_new_isnotprior.groupby('HS2').agg(
    HS2_F = ('HS2_F','first'),
    Import_R = ('Value_R','sum'),
    Import_D = ('Value_D','sum'),

    Hoghough_Voroudi_R = ('HSum_R','sum')
).reset_index()

HS2s['Darsad_Varedat']=round(100*HS2s['Import_R']/HS2s['Import_R'].sum(),3)
HS2s['H_Perc'] = round(100*HS2s['Hoghough_Voroudi_R']/HS2s['Import_R'],3)

HS2s.to_excel(f'C:/TradeGroup/Reports/Hoghough_Voroudi_isnotPrior_HS2.xlsx',index=False)



```
