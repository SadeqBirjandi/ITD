```python
mask1397_New = Nerkh_Tarefe_1397['Tcode'] != 'جمع کل'

mask1398_New = Nerkh_Tarefe_1398['Tcode'] != 'جمع کل'

mask1399_New = Nerkh_Tarefe_1399['Tcode'] != 'جمع کل'

mask1400_New = Nerkh_Tarefe_1400['Tcode'] != 'جمع کل'

mask1401_New = Nerkh_Tarefe_1401['Tcode'] != 'جمع کل'

mask1402_New = Nerkh_Tarefe_1402['Tcode'] != 'جمع کل'


filtered_1397_New = Nerkh_Tarefe_1397[mask1397_New]

filtered_1398_New = Nerkh_Tarefe_1398[mask1398_New]

filtered_1399_New = Nerkh_Tarefe_1399[mask1399_New]

filtered_1400_New = Nerkh_Tarefe_1400[mask1400_New]

filtered_1401_New = Nerkh_Tarefe_1401[mask1401_New]

filtered_1402_New = Nerkh_Tarefe_1402[mask1402_New]


filtered_1397_New['Year']=1397
filtered_1398_New['Year']=1398
filtered_1399_New['Year']=1399
filtered_1400_New['Year']=1400
filtered_1401_New['Year']=1401
filtered_1402_New['Year']=1402
```

```python
All_Totals_New = pd.concat([filtered_1397_New,filtered_1398_New,filtered_1399_New,filtered_1400_New,filtered_1401_New,filtered_1402_New])


conditions = [
    (All_Totals_New['HPerc_R'] == 0),             
    ((All_Totals_New['HPerc_R'] > 0) & (All_Totals_New['HPerc_R'] <= 5)),
    ((All_Totals_New['HPerc_R'] > 5) & (All_Totals_New['HPerc_R'] <= 10)),
    ((All_Totals_New['HPerc_R'] > 10) & (All_Totals_New['HPerc_R'] <= 20)),
    ((All_Totals_New['HPerc_R'] > 20) & (All_Totals_New['HPerc_R'] <= 50)),
    ((All_Totals_New['HPerc_R'] > 50) & (All_Totals_New['HPerc_R'] <= 100)),
    ((All_Totals_New['HPerc_R'] > 100))
]

choices = [
    'Zero',
    'from 0 to 5',
    'from 5 to 10',
    'from 10 to 20', 
    'from 20 to 50',   
    'from 50 to 100',
    'more than 100'
]

# Use np.select() to generate the new column 'B'
All_Totals_New['HPerc_Cat'] = np.select(conditions, choices, default='other')


```

```python
All_Totals_New['HPerc_Cat'].unique()
```

```python
By_HPerc_Cat = All_Totals_New.groupby(['Year','HPerc_Cat']).agg(
    Value_R = ('Value_R','sum'),
    Value_D = ('Value_D','sum'),
    HSum_R = ('HSum_R','sum'),
    Codes_Count = ('Tcode','count')
    
    
).reset_index()

#By_HPerc_Cat['Sum_D_Year']=By_HPerc_Cat.groupby('Year').sum('Value_D')
By_HPerc_Cat['Sum_D_Year'] = By_HPerc_Cat.groupby('Year')['Value_D'].transform('sum')
By_HPerc_Cat['Sum_R_Year'] = By_HPerc_Cat.groupby('Year')['Value_R'].transform('sum')

By_HPerc_Cat['ShareValue_D_Year'] = By_HPerc_Cat['Value_D']/By_HPerc_Cat['Sum_D_Year']

By_HPerc_Cat['ShareValue_R_Year'] = By_HPerc_Cat['Value_R']/By_HPerc_Cat['Sum_R_Year']






By_HPerc_Cat['Sum_H_R_Year'] = By_HPerc_Cat.groupby('Year')['HSum_R'].transform('sum')


By_HPerc_Cat['Shared_H_R_Year'] = By_HPerc_Cat['HSum_R']/By_HPerc_Cat['Sum_H_R_Year']


By_HPerc_Cat['Sum_N_Year'] = By_HPerc_Cat.groupby('Year')['Codes_Count'].transform('sum')

By_HPerc_Cat['Share_of_Codes'] = By_HPerc_Cat['Codes_Count']/By_HPerc_Cat['Sum_N_Year']



By_HPerc_Cat


By_HPerc_Cat.to_excel('C:/TradeGroup/Reports/3 ماه اول/Tarefe_By_Cat.xlsx', index=False)
```
