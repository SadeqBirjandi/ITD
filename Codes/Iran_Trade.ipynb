```python
import pandas as pd
import numpy as np
import os
```

```python
os.chdir("C:\TradeGroup")
my_wd = os.getcwd()
new_directory = my_wd
```

```python
year = 1403  
for year in range(1403,1404):
    exports_file_path = f"{my_wd}/Data/Cleaned_Data/Y{year}Exports.xlsx"
    imports_file_path = f"{my_wd}/Data/Cleaned_Data/Y{year}Imports.xlsx"
    #if year!=1394:
    Exports = pd.read_excel(exports_file_path)
    Imports = pd.read_excel(imports_file_path)
    
    

    if "Year" not in Exports.columns:
        Exports['Year']= year
    if "Year" not in Imports.columns:
        Imports['Year']= year    

    if 'Month' not in Exports.columns:
        Exports['Month']= "Not_Available"
    if 'Month' not in Imports.columns:
        Imports['Month']= "Not_Available"



    if 'Custom' not in Exports.columns:
        Exports['Custom']= "Not_Available"
    if 'Custom' not in Imports.columns:
        Imports['Custom']= "Not_Available"
    

    Exports_mod1 = Exports.groupby(['Year','Tcode','Country','Custom','Month']).agg(
        Tdesc = ('Tdesc','first'),
        Dollars = ("Dollars",'sum'),
        Rials = ('Rials','sum'),
        Weight = ('Weight','sum')
    ).reset_index()


    Imports_mod1 = Imports.groupby(['Year','Tcode','Country','Custom','Month']).agg(
        Tdesc = ('Tdesc','first'),
        Dollars = ("Dollars",'sum'),
        Rials = ('Rials','sum'),
        Weight = ('Weight','sum')
    ).reset_index()

    Imports_mod1['Trade'] = "Import"
    Exports_mod1['Trade'] = "Export"

    Imports_mod1['Tcode'] = Imports_mod1['Tcode'].astype('int')
    Imports_mod1['HS6']=Imports_mod1['Tcode'].astype('str').str.slice(0,-2)

    Exports_mod1['Tcode'] = Exports_mod1['Tcode'].astype('int')
    Exports_mod1['HS6']=Exports_mod1['Tcode'].astype('str').str.slice(0,-2)

    Iran_Trade= pd.concat([Imports_mod1,Exports_mod1])

    if year == 1403:

        def get_nonzero_tdesc(tdesc, tcode):
            if tdesc == 0:
                nonzero_tdesc_rows = Iran_Trade.loc[(Iran_Trade['Tcode'] == tcode) & (Iran_Trade['Tdesc'] != 0)]
                if not nonzero_tdesc_rows.empty:
                    first_nonzero_tdesc = nonzero_tdesc_rows.iloc[0]['Tdesc']
                    return first_nonzero_tdesc
                else:
                    # If no non-zero Tdesc value is found for this Tcode, return the original value
                    return tdesc
            else:
                return tdesc

        # Apply the function to the Tdesc column
        Iran_Trade['Tdesc'] = Iran_Trade.apply(lambda row: get_nonzero_tdesc(row['Tdesc'], row['Tcode']), axis=1)


    Iran_Trade.to_excel(f'C:/TradeGroup/Data/Dashboard/IranTrade{year}.xlsx')
```
