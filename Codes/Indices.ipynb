```python
import pandas as pd
import numpy as np
import os
```

```python
# Set the working directory
os.chdir("C:\TradeGroup")
my_wd = os.getcwd()
new_directory = my_wd

def create_folder(folder_path):
    if not os.path.exists(f'{new_directory}/{folder_path}'):
        os.makedirs(f'{new_directory}/{folder_path}')
    else:
        print(f'Already Exists')

#create_folder('Data/Indices')
```

```python
create_folder('Data/Indices/Variety_Consetration')
baseyear=1380
for year in range(1380,1403):

        try:
                del Exports
        except NameError:
                pass

        try:
                del Imports
        except NameError:
                pass

        # Load data
        exports_path = os.path.join(my_wd, f"Data/Cleaned_Data/Y{year}Exports.xlsx")
        imports_path = os.path.join(my_wd, f"Data/Cleaned_Data/Y{year}Imports.xlsx")


        Exports = pd.read_excel(exports_path)
        # Imports = pd.read_excel(imports_path)

        # Imports.dropna(subset=['Tcode'],inplace=True)
        Exports.dropna(subset=['Tcode'],inplace=True)

        
        # Imports['Dollars']=Imports['Dollars'].astype(float)
        # Imports['Weight']=Imports['Weight'].astype(float)

        Exports['Dollars']=Exports['Dollars'].astype(float)
        Exports['Weight']=Exports['Weight'].astype(float) 

        Exports['HS6']=Exports['Tcode']//100
        #Exports['HS6']=Exports['HS6'].astype('str')
        Sum_Exports=Exports['Dollars'].sum()

        Exports_by_HS6=Exports.groupby('HS6').agg(
        Dollars_Tot=('Dollars','sum')
        ).reset_index()

        Exports_by_HS6[f'HS_Counts_{year}']=Exports_by_HS6['HS6'].nunique()

        Exports_by_HS6[f'Share_{year}']=Exports_by_HS6['Dollars_Tot']/Sum_Exports

        Exports_by_HS6[f'Share_Squared_{year}']=Exports_by_HS6[f'Share_{year}']**2

        Var_Cons=pd.DataFrame(data={'Year':[year],'Variety': Exports_by_HS6[f'HS_Counts_{year}'].unique(),'Consentration':[Exports_by_HS6[f'Share_Squared_{year}'].sum()]})
        if year==baseyear:
                Var_Cons_Series=Var_Cons
        else:
                Var_Cons_Series = pd.concat([Var_Cons_Series,Var_Cons])        





print(year)

```
