```python
Import_03 = pd.read_excel(f'C:/TradeGroup/Data/RawData/Imp140303first.xlsx')
Import_03.columns=['Year','Custom','Country','HSCode','Tdesc','Weight_03','Rials_03','Dollars_03']


Export_03 = pd.read_excel(f'C:/TradeGroup/Data/RawData/Exp140303first.xlsx')
Export_03.columns=['Year','Custom','Country','HSCode','Tdesc','Weight_03','Rials_03','Dollars_03']
```

```python
Import_06 = Import_06[Import_06['Year']!="جمع"]
Export_06 = Export_06.dropna(axis=0)

Export_06


Import_03 = Import_03[Import_03['Year']!="جمع"]
Export_03 = Export_03.dropna(axis=0)

Export_03
```

```python
CCI_Codes = pd.read_excel(f"C:/TradeGroup/Reports/CCI_Codes_HS8.xlsx")
```

```python
Imports_06_WithCCI = Import_06.merge(CCI_Codes,on="HSCode",how="left")
Exports_06_WithCCI = Export_06.merge(CCI_Codes,on="HSCode",how="left")


Imports_03_WithCCI = Import_03.merge(CCI_Codes,on="HSCode",how="left")
Exports_03_WithCCI = Export_03.merge(CCI_Codes,on="HSCode",how="left")

Exports_03_WithCCI
```

```python
Imports_06_WithCCI_Sum = Imports_06_WithCCI.groupby("CCI").agg(
    Dollars_6 = ("Dollars_06","sum"),
    Weight_6 = ("Weight_06", "sum"),
    Rials_6 = ("Rials_06","sum")
).reset_index()



Imports_01_WithCCI_Sum = Imports_03_WithCCI.groupby("CCI").agg(
    Dollars_1403_01 = ("Dollars_03","sum"),
    Weight_1403_01 = ("Weight_03", "sum"),
    Rials_1403_01 = ("Rials_03","sum")
).reset_index()



Imports_0102 =Imports_01_WithCCI_Sum.merge(Imports_06_WithCCI_Sum, on="CCI", how="outer")

Imports_0102['Dollars_1403_02'] = Imports_0102['Dollars_6']-Imports_0102['Dollars_1403_01']
Imports_0102['Weight_1403_02'] = Imports_0102['Weight_6']-Imports_0102['Weight_1403_01']
Imports_0102['Rials_1403_02'] = Imports_0102['Rials_6']-Imports_0102['Rials_1403_01']


#Imports_0102.to_excel(f"C:/TradeGroup/Reports/CCI_140302.xlsx")

Imports_0102
```

```python
Import_1402 = pd.read_excel(f'C:/TradeGroup/Data/Cleaned_Data/Y1402Imports.xlsx')
```

```python
Import_1402.rename(columns={"Tcode":"HSCode"},inplace=True)
```

```python
Imports_1402_WithCCI = Import_1402.merge(CCI_Codes,on="HSCode",how="left")

Imports_1402_WithCCI['Quarter'] = ((Imports_1402_WithCCI['Month']-1)//3+1).astype("int")

Imports_1402_WithCCI
```

```python
Imports_1402_WithCCI_Sum_01 = Imports_1402_WithCCI[Imports_1402_WithCCI['Quarter']==1].groupby("CCI").agg(
    Dollars_01_1402 = ("Dollars","sum"),
    Weight_01_1402 = ("Weight", "sum"),
    Rials_01_1402 = ("Rials","sum")
).reset_index()



Imports_1402_WithCCI_Sum_02 = Imports_1402_WithCCI[Imports_1402_WithCCI['Quarter']==2].groupby("CCI").agg(
    Dollars_02_1402 = ("Dollars","sum"),
    Weight_02_1402 = ("Weight", "sum"),
    Rials_02_1402 = ("Rials","sum")
).reset_index()



Imports_1402_WithCCI_Sum_03 = Imports_1402_WithCCI[Imports_1402_WithCCI['Quarter']==3].groupby("CCI").agg(
    Dollars_03_1402 = ("Dollars","sum"),
    Weight_03_1402 = ("Weight", "sum"),
    Rials_03_1402 = ("Rials","sum")
).reset_index()



Imports_1402_WithCCI_Sum_04 = Imports_1402_WithCCI[Imports_1402_WithCCI['Quarter']==4].groupby("CCI").agg(
    Dollars_04_1402 = ("Dollars","sum"),
    Weight_04_1402 = ("Weight", "sum"),
    Rials_04_1402 = ("Rials","sum")
).reset_index()


Imports_1402 = Imports_1402_WithCCI_Sum_01.merge(Imports_1402_WithCCI_Sum_02,on="CCI")
Imports_1402 = Imports_1402.merge(Imports_1402_WithCCI_Sum_03,on="CCI")
Imports_1402 = Imports_1402.merge(Imports_1402_WithCCI_Sum_04,on="CCI")

Imports_1402
```

```python
Imports_1402_1403 = Imports_1402.merge(Imports_0102,on="CCI")

Imports_1402_1403.to_excel(f"C:/TradeGroup/Reports/CCI_1402_140302.xlsx")

```
