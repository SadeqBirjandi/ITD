```python
month = 4
last_month = 3
Imports_Synthesized = pd.read_excel(f'C:/TradeGroup/Data/RawData/Imp1403first{last_month}_synthetic.xlsx')
Exports_Synthesized = pd.read_excel(f'C:/TradeGroup/Data/RawData/Exp1403first{last_month}_synthetic.xlsx')


Imports_last_month = pd.read_excel(f'C:/TradeGroup/Data/RawData/Imp14030{last_month}first.xlsx')
Exports_last_month = pd.read_excel(f'C:/TradeGroup/Data/RawData/Exp14030{last_month}first.xlsx')


Imports_month = pd.read_excel(f'C:/TradeGroup/Data/RawData/Imp14030{month}first.xlsx')
Exports_month = pd.read_excel(f'C:/TradeGroup/Data/RawData/Exp14030{month}first.xlsx')



```

```python
Imports_last_month.columns=['Year','Custom','Country','Tcode','Tdesc','Weight_last_month','Rials_last_month','Dollars_last_month']
Exports_last_month.columns=['Year','Custom','Country','Tcode','Tdesc','Weight_last_month','Rials_last_month','Dollars_last_month']

Imports_month.columns=['Year','Custom','Country','Tcode','Tdesc','Weight','Rials','Dollars']
Exports_month.columns=['Year','Custom','Country','Tcode','Tdesc','Weight','Rials','Dollars']


Imports_month=Imports_month.groupby(['Tcode','Country','Custom']).agg(
    Weight_last_month_month = ('Weight','sum'),
    Rials_last_month_month = ('Rials','sum'),
    Dollars_last_month_month = ('Dollars','sum')
).reset_index()


Exports_month=Exports_month.groupby(['Tcode','Country','Custom']).agg(
    Weight_last_month_month = ('Weight','sum'),
    Rials_last_month_month = ('Rials','sum'),
    Dollars_last_month_month = ('Dollars','sum')
).reset_index()



Imports_last_month=Imports_last_month.groupby(['Tcode','Country','Custom']).agg(
    Tdesc = ('Tdesc','first'),
    Weight_last_month = ('Weight_last_month','sum'),
    Rials_last_month = ('Rials_last_month','sum'),
    Dollars_last_month = ('Dollars_last_month','sum')
).reset_index()


Exports_last_month=Exports_last_month.groupby(['Tcode','Country','Custom']).agg(
    Tdesc = ('Tdesc','first'),
    Weight_last_month = ('Weight_last_month','sum'),
    Rials_last_month = ('Rials_last_month','sum'),
    Dollars_last_month = ('Dollars_last_month','sum')
).reset_index()


```

```python
Imports_last_month['Tcode']=Imports_last_month['Tcode'].astype('int')
Exports_last_month['Tcode']=Exports_last_month['Tcode'].astype('int')

Imports_month['Tcode']=Imports_month['Tcode'].astype('int')
Exports_month['Tcode']=Exports_month['Tcode'].astype('int')


Imports_tillthismonth = Imports_month.merge(Imports_last_month,on=['Tcode','Country','Custom'],how='outer')
Exports_tillthismonth = Exports_month.merge(Exports_last_month,on=['Tcode','Country','Custom'],how='outer')

Imports_tillthismonth = Imports_tillthismonth.dropna(axis=0,subset=['Country'])
Exports_tillthismonth = Exports_tillthismonth.dropna(axis=0,subset=['Country'])

Imports_tillthismonth.fillna(0,inplace=True)
Exports_tillthismonth.fillna(0,inplace=True)
```

```python
Exports_tillthismonth[f'Weight_0{month}']=Exports_tillthismonth['Weight_last_month_month']- Exports_tillthismonth['Weight_last_month']
Exports_tillthismonth[f'Dollars_0{month}']=Exports_tillthismonth['Dollars_last_month_month']- Exports_tillthismonth['Dollars_last_month']
Exports_tillthismonth[f'Rials_0{month}']=Exports_tillthismonth['Rials_last_month_month']- Exports_tillthismonth['Rials_last_month']


Imports_tillthismonth[f'Weight_0{month}']=Imports_tillthismonth['Weight_last_month_month']- Imports_tillthismonth['Weight_last_month']
Imports_tillthismonth[f'Dollars_0{month}']=Imports_tillthismonth['Dollars_last_month_month']- Imports_tillthismonth['Dollars_last_month']
Imports_tillthismonth[f'Rials_0{month}']=Imports_tillthismonth['Rials_last_month_month']- Imports_tillthismonth['Rials_last_month']
```

```python
Exports_justthismonth=Exports_tillthismonth[['Custom','Country','Tcode','Tdesc',f'Weight_0{month}',f'Rials_0{month}',f'Dollars_0{month}']]

Imports_justthismonth=Imports_tillthismonth[['Custom','Country','Tcode','Tdesc',f'Weight_0{month}',f'Rials_0{month}',f'Dollars_0{month}']]
```

```python
Exports_justthismonth=Exports_justthismonth[Exports_justthismonth.Dollars_04!=0]
Exports_justthismonth['Month']=month
Exports_justthismonth['Year']=1403



Imports_justthismonth=Imports_justthismonth[Imports_justthismonth.Dollars_04!=0]
Imports_justthismonth['Month']=month
Imports_justthismonth['Year']=1403
```

```python
Exports_justthismonth=Exports_justthismonth.rename(columns={f'Weight_0{month}':'Weight',f'Rials_0{month}':'Rials',f'Dollars_0{month}':'Dollars'})

Imports_justthismonth=Imports_justthismonth.rename(columns={f'Weight_0{month}':'Weight',f'Rials_0{month}':'Rials',f'Dollars_0{month}':'Dollars'})
```
