```python
year = 1400
Trade_Series = pd.DataFrame()
for year in range (1390,1404):    

    Imports = pd.read_excel(f'C:/TradeGroup/Data/Cleaned_Data/Y{year}Imports.xlsx')
    Exports = pd.read_excel(f'C:/TradeGroup/Data/Cleaned_Data/Y{year}Exports.xlsx')         

    Exports_Dollars = Exports['Dollars'].sum()
    Imports_Dollars = Imports['Dollars'].sum()
    Exports_Weight = Exports['Weight'].sum()
    Imports_Weight = Imports['Weight'].sum()
    



    

    Trade = pd.DataFrame({"Year":year,
                        "Exports":Exports_Dollars,
                        "Imports": Imports_Dollars,
                        "Trade_Volume": Exports_Dollars+Imports_Dollars,
                        "Balance": Exports_Dollars-Imports_Dollars,
                        "Exports_Weight":Exports_Weight,
                        "Imports_Weigh":Imports_Weight },index=[year]
                        )

    Trade_Series = pd.concat([Trade_Series,Trade])



Trade_Series['Exports_Change']= Trade_Series['Exports'].pct_change()
Trade_Series['Imports_Change']= Trade_Series['Imports'].pct_change()
Trade_Series['Trade_Volume_Change'] = Trade_Series['Trade_Volume'].pct_change()
Trade_Series['Balance_Change'] = Trade_Series['Balance'].pct_change()


Trade_Series.to_excel(f"C:/TradeGroup/Reports/Trade_Report/ExpImp_Value_Price.xlsx")
```
