```python
import pandas as pd
import numpy as np
import os
```

```python
os.chdir("C:\TradeGroup")
my_wd = os.getcwd()
new_directory = my_wd
```

```python
Iran_to_Indonasia=pd.read_excel(f"C:/TradeGroup/Reports/Iran-Indonasia-Be.xlsx")
Iran_to_Indonasia['شماره تعرفه'].fillna(0,inplace=True)
Iran_to_Indonasia['HS8']=Iran_to_Indonasia['شماره تعرفه'].astype(int)
Iran_Exports_Codes = pd.unique(Iran_to_Indonasia['HS8'])
Iran_Exports_Codes

My_Excel_File=pd.ExcelFile(f"C:/TradeGroup/Reports/Iran-Indonasia-Alef.xlsx")
dfs = []
for sheet_name in My_Excel_File.sheet_names:
    df= My_Excel_File.parse(sheet_name)
    dfs.append(df)

The_Commodities = pd.concat(dfs,ignore_index=True)
The_Commodities.rename(columns={'شماره تعرفه':"HS8"},inplace=True)
The_Commodities['HS8']=pd.to_numeric(The_Commodities['HS8'])
Iran_Imports_Codes = pd.unique(The_Commodities['HS8'])
```

```python
year = 1398
baseyear = 1398
for year in range(1398,1403):
    exports_file_path = f"{my_wd}/Data/Cleaned_Data/Y{year}Exports.xlsx"

    Exports = pd.read_excel(exports_file_path)


    Exports_HSs = Exports.groupby('Tcode').agg(
        Dollars = ('Dollars','sum')
    ).reset_index()

    Exports_HSs.columns=Exports_HSs.columns.str.replace("Dollars",f'Dollars_{year}')

    Exports_HSs['Tarjihi']=np.where(np.isin(Exports_HSs['Tcode'], Iran_Exports_Codes),1,0 )


    Exports_HSs[f'rank_{year}'] = Exports_HSs[f'Dollars_{year}'].rank(ascending= False)


    if year == baseyear:
        Exports_HSs_Series = Exports_HSs
    else:
        Exports_HSs_Series = Exports_HSs_Series.merge(Exports_HSs, on=['Tcode','Tarjihi'], how='outer')
        Exports_HSs_Series.fillna(0, inplace= True) 


Exports_HSs_Series
```

```python
columns = [f'Dollars_{year}' for year in range(1398,1403)]

Exports_HSs_Series[f'Total_Dollars']=Exports_HSs_Series[columns].sum(axis=1)

Exports_HSs_Series['Share_Total']= Exports_HSs_Series['Total_Dollars']/Exports_HSs_Series['Total_Dollars'].sum()

Exports_HSs_Series['Rank'] = Exports_HSs_Series['Total_Dollars'].rank(ascending= False)

Exports_HSs_Series
```

```python
Exports_uniques = Exports[['Tcode','Tdesc']]
Exports_uniques = Exports_uniques.drop_duplicates().dropna()

Exports_uniques
```

```python
Exports_HSs_Series_Tdesc = Exports_uniques.merge(Exports_HSs_Series, on='Tcode', how = 'right')

Exports_HSs_Series_Tdesc.to_excel(f"C:/TradeGroup/Reports/Iran_Exports_Ranking.xlsx", index= False)
```
