```python
import os
import glob
import pandas as pd

# Target folder path
folder_path = r"C:/TradeGroup/Reports/All_Countries"

# Get a list of Excel file paths in the target folder
excel_files = glob.glob(os.path.join(folder_path, "*.xlsx"))
country_names = [os.path.splitext(os.path.basename(file_path))[0].replace("_Data", "") for file_path in excel_files]
```

```python
os.chdir("C:\TradeGroup")
my_wd = os.getcwd()
new_directory = my_wd

def create_folder(folder_path):
    if not os.path.exists(f'{new_directory}/{folder_path}'):
        os.makedirs(f'{new_directory}/{folder_path}')
    else:
        print(f'Already Exists')

create_folder('Data/Indices')
```

```python
Iran_Exports = pd.read_excel(f'{my_wd}/Data/Indices/Trademap_IranExports_2022.xlsx')



Iran_Exports['Code'] = Iran_Exports['Code'].str.replace("'","").astype('int')
#Iran_Exports['Code'] = Iran_Exports['Code'].astype("int")

remaining_columns = Iran_Exports.columns[2:]


remaining_columns=remaining_columns.str.replace("Exported value in ","")

Iran_Exports.columns=Iran_Exports.columns.str.replace("Exported value in ","")


Iran_Exports = Iran_Exports[Iran_Exports['Code']!=999999]

Iran_Exports.columns = Iran_Exports.columns[0:2].tolist() + [col + '_Iran_Exports' for col in Iran_Exports.columns[2:]]

for col in Iran_Exports.columns[2:]:
    Iran_Exports[f'{col}_Share'] = Iran_Exports[col]/(Iran_Exports[col].sum())

```

```python
for country in country_names:
    print(country)
    df = pd.read_excel(f'{folder_path}/{country}_Data.xlsx')

    country = country.replace("_"," ")


    df = df[df['Product code']!= 999999]
    for col in df.columns[2:]:
        df[f'{col}_Share'] = df[col]/(df[col].sum())

    df = df.fillna(0)

    df['Code'] = df['Product code']

    country_df= pd.DataFrame({'Year':range(2004,2024),f'{country}':0})
    country_df[f'{country}']= country_df[f'{country}'].astype('float')

    for year in range(2003,2023):
        if (f"Imports_from_World_{year}_Share" in df.columns) & (f"{year}_Iran_Exports_Share" in Iran_Exports.columns):
            merged_data = df[['Code',f"Imports_from_World_{year}_Share"]].merge(Iran_Exports[['Code',f"{year}_Iran_Exports_Share"]])
            merged_data.fillna(0,inplace=True)
            merged_data[f'Comp_Good_{year}']=abs(merged_data[f"Imports_from_World_{year}_Share"]-merged_data[f"{year}_Iran_Exports_Share"])/2

            Year_Value = 100*(1-(merged_data[f'Comp_Good_{year}'].sum()))

            country_df.loc[country_df['Year']==year,f'{country}']=Year_Value


    if country=="Afghanistan":
                        country_df_all = country_df
    else:
                        country_df_all = country_df_all.merge(country_df,on="Year")

            
        

```

```python
country_df_all.to_excel(f'{my_wd}\Data\Indices\Iran_TCI_Final_2003to2022.xlsx',index=False)

country_df_all
```
