```python
import pandas as pd
import numpy as np
import os
```

```python
os.chdir("C:\TradeGroup")
my_wd = os.getcwd()
new_directory = my_wd
```

```python
Countries = ['Armenia','Belarus','Kyrgyzstan','Kazakhstan','Russian Federation']

for country in Countries:
    Imports = pd.read_excel(f'C:/TradeGroup/Reports/Eurasia/گزارش/{country}_Imports_Product.xlsx')
    Exports = pd.read_excel(f'C:/TradeGroup/Reports/Eurasia/گزارش/{country}_Exports_Product.xlsx')

    Imports = Imports[['Code','Product label' , 'Y2023']]
    Exports = Exports[['Code','Product label' , 'Y2023']]


    Imports["Rank"] = Imports["Y2023"].rank(ascending=False)
    Exports["Rank"] = Exports["Y2023"].rank(ascending=False)
    

    Imports_First_10 = Imports[Imports['Rank']<11]
    Exports_First_10 = Exports[Exports['Rank']<11]

    Import_Sum_First_10 = pd.DataFrame({'Code':999999,'Product label':"First 10 Sum",'Y2023': Imports_First_10['Y2023'].sum(),'Rank':11},index=[100])
    Export_Sum_First_10 = pd.DataFrame({'Code':999999,'Product label':"First 10 Sum",'Y2023': Exports_First_10['Y2023'].sum(),'Rank':11},index=[100])


    Import_Sum_Other = pd.DataFrame({'Code':9999999,'Product label':"Others Sum",'Y2023': Imports.loc[Imports['Rank']>10,'Y2023'].sum(),'Rank':12},index=[101])
    Export_Sum_Other = pd.DataFrame({'Code':9999999,'Product label':"Others Sum",'Y2023': Exports.loc[Exports['Rank']>10,'Y2023'].sum(),'Rank':12},index=[101])


    Imports_New = pd.concat([Imports_First_10,Import_Sum_First_10])  
    Exports_New = pd.concat([Exports_First_10,Export_Sum_First_10])

    Imports_New_New = pd.concat([Imports_New,Import_Sum_Other])
    Exports_New_New = pd.concat([Exports_New,Export_Sum_Other])

    Imports_New_New.to_excel(f'C:/TradeGroup/Reports/Eurasia/گزارش/{country}_Imports_Product_First10.xlsx')
    Exports_New_New.to_excel(f'C:/TradeGroup/Reports/Eurasia/گزارش/{country}_Exports_Product_First10.xlsx')




    





```
