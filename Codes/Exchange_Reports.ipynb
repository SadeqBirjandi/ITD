```python
Exchange_Merged = Exchange.merge(HSs_Desc,on=['HS6','HS5'],how='left')
Exchange_Merged_NAs = Exchange_Merged[Exchange_Merged['HS4'].isnull()]
Exchange_Merged_NonNAs = Exchange_Merged[~Exchange_Merged['HS4'].isnull()]

Exchange_Merged_NAs_tobemerged = Exchange[Exchange['Tcode'].isin(Exchange_Merged_NAs['Tcode'])]
```

```python
Exchange_concated_final['Year']=Exchange_concated_final['Full_Date'].str.slice(0,4).astype('int')
Exchange_concated_final['Month']=Exchange_concated_final['Full_Date'].str.slice(5,7).astype('int')
Exchange_concated_final['Day']=Exchange_concated_final['Full_Date'].str.slice(9,10).astype('int')
```

```python
Exchange_concated_final.loc[(Exchange_concated_final['HS6']==999999) |
                                                   ((Exchange_concated_final['HS6']>0)&(Exchange_concated_final['HS6']<990000)),'HS_Type']="Good"


Exchange_concated_final.loc[((Exchange_concated_final['HS6']<999999)&(Exchange_concated_final['HS6']>990000)),'HS_Type']="Service"

Exchange_concated_final.loc[(Exchange_concated_final['HS6']==0),'HS_Type']="Other"

Exchange_new = Exchange_concated_final
```

```python
Takhsis_Type=Exchange_concated_final.groupby(['Year','Month',' Takhsis_Type']).agg(
    MDollars = ('MDollars','sum')
).reset_index()



Month=Exchange_concated_final.groupby(['Year','Month']).agg(
    MDollars = ('MDollars','sum')
).reset_index()

HS_Type=Exchange_concated_final.groupby('HS_Type').agg(
    MDollars = ('MDollars','sum')
).reset_index()


HS_Type_Month=Exchange_concated_final.groupby(['Year','Month','HS_Type']).agg(

    MDollars = ('MDollars','sum')
).reset_index()



HS2_Total=Exchange_concated_final.groupby(['HS_Type','HS2']).agg(
    HS2_F = ('HS2_F','first'),
    HS2_E = ('HS2_E','first'),
    MDollars = ('MDollars','sum')
).reset_index()

HS2_Year=Exchange_concated_final.groupby(['Year','HS_Type','HS2',]).agg(
    HS2_F = ('HS2_F','first'),
    HS2_E = ('HS2_E','first'),
    MDollars = ('MDollars','sum')
).reset_index()



HS2_Month=Exchange_concated_final.groupby(['Year','Month','HS_Type','HS2']).agg(
    HS2_F = ('HS2_F','first'),
    HS2_E = ('HS2_E','first'),
    MDollars = ('MDollars','sum')
).reset_index()






HS4_Total=Exchange_concated_final.groupby(['HS_Type','HS4']).agg(
    HS2 = ('HS2','first'),
    HS2_F = ('HS2_F','first'),
    HS2_E = ('HS2_E','first'),
    HS4_F = ('HS4_F','first'),
    HS4_E = ('HS4_E','first'),
    MDollars = ('MDollars','sum')
).reset_index()

HS4_Year=Exchange_concated_final.groupby(['Year','HS_Type','HS4']).agg(
    HS2 = ('HS2','first'),
    HS2_F = ('HS2_F','first'),
    HS2_E = ('HS2_E','first'),
    HS4_F = ('HS4_F','first'),
    HS4_E = ('HS4_E','first'),
    MDollars = ('MDollars','sum')
).reset_index()



HS4_Month=Exchange_concated_final.groupby(['Year','Month','HS_Type','HS4']).agg(
    HS2 = ('HS2','first'),
    HS2_F = ('HS2_F','first'),
    HS2_E = ('HS2_E','first'),
    HS4_F = ('HS4_F','first'),
    HS4_E = ('HS4_E','first'),
    MDollars = ('MDollars','sum')
).reset_index()






HS6_Total=Exchange_concated_final.groupby(['HS_Type','HS6']).agg(
    HS2 = ('HS2','first'),
    HS2_F = ('HS2_F','first'),
    HS2_E = ('HS2_E','first'),
    HS4 = ('HS4','first'),
    HS4_F = ('HS4_F','first'),
    HS4_E = ('HS4_E','first'),
    HS6_F = ('HS6_F','first'),
    HS6_E = ('HS6_E','first'),
    MDollars = ('MDollars','sum')
).reset_index()

HS6_Year=Exchange_concated_final.groupby(['Year','HS_Type','HS6']).agg(
    HS2 = ('HS2','first'),
    HS2_F = ('HS2_F','first'),
    HS2_E = ('HS2_E','first'),
    HS4 = ('HS4','first'),
    HS4_F = ('HS4_F','first'),
    HS4_E = ('HS4_E','first'),
    HS6_F = ('HS6_F','first'),
    HS6_E = ('HS6_E','first'),
    MDollars = ('MDollars','sum')
).reset_index()



HS6_Month=Exchange_concated_final.groupby(['Year','Month','HS_Type','HS6']).agg(
    HS2 = ('HS2','first'),
    HS2_F = ('HS2_F','first'),
    HS2_E = ('HS2_E','first'),
    HS4 = ('HS4','first'),
    HS4_F = ('HS4_F','first'),
    HS4_E = ('HS4_E','first'),
    HS6_F = ('HS6_F','first'),
    HS6_E = ('HS6_E','first'),
    MDollars = ('MDollars','sum')
).reset_index()





HS8_Total=Exchange_concated_final.groupby(['HS_Type','Tcode']).agg(
    HS2 = ('HS2','first'),
    HS2_F = ('HS2_F','first'),
    HS2_E = ('HS2_E','first'),
    HS4 = ('HS4','first'),
    HS4_F = ('HS4_F','first'),
    HS4_E = ('HS4_E','first'),
    HS6 = ('HS6','first'),
    HS6_F = ('HS6_F','first'),
    HS6_E = ('HS6_E','first'),
    HS8_F = ('Tdesc','first'),
    MDollars = ('MDollars','sum')
).reset_index()

HS8_Year=Exchange_concated_final.groupby(['Year','HS_Type','Tcode']).agg(
    HS2 = ('HS2','first'),
    HS2_F = ('HS2_F','first'),
    HS2_E = ('HS2_E','first'),
    HS4 = ('HS4','first'),
    HS4_F = ('HS4_F','first'),
    HS4_E = ('HS4_E','first'),
    HS6 = ('HS6','first'),
    HS6_F = ('HS6_F','first'),
    HS6_E = ('HS6_E','first'),
    HS8_F = ('Tdesc','first'),
    MDollars = ('MDollars','sum')
).reset_index()



HS8_Month=Exchange_concated_final.groupby(['Year','Month','HS_Type','Tcode']).agg(
    HS2 = ('HS2','first'),
    HS2_F = ('HS2_F','first'),
    HS2_E = ('HS2_E','first'),
    HS4 = ('HS4','first'),
    HS4_F = ('HS4_F','first'),
    HS4_E = ('HS4_E','first'),
    HS6 = ('HS6','first'),
    HS6_F = ('HS6_F','first'),
    HS6_E = ('HS6_E','first'),
    HS8_F = ('Tdesc','first'),
    MDollars = ('MDollars','sum')
).reset_index()






HS8_Total_Takh=Exchange_concated_final.groupby(['HS_Type','Tcode',' Takhsis_Type']).agg(
    HS2 = ('HS2','first'),
    HS2_F = ('HS2_F','first'),
    HS2_E = ('HS2_E','first'),
    HS4 = ('HS4','first'),
    HS4_F = ('HS4_F','first'),
    HS4_E = ('HS4_E','first'),
    HS6 = ('HS6','first'),
    HS6_F = ('HS6_F','first'),
    HS6_E = ('HS6_E','first'),
    HS8_F = ('Tdesc','first'),
    MDollars = ('MDollars','sum')
).reset_index()

HS8_Year_Takh=Exchange_concated_final.groupby(['Year','HS_Type','Tcode',' Takhsis_Type']).agg(
    HS2 = ('HS2','first'),
    HS2_F = ('HS2_F','first'),
    HS2_E = ('HS2_E','first'),
    HS4 = ('HS4','first'),
    HS4_F = ('HS4_F','first'),
    HS4_E = ('HS4_E','first'),
    HS6 = ('HS6','first'),
    HS6_F = ('HS6_F','first'),
    HS6_E = ('HS6_E','first'),
    HS8_F = ('Tdesc','first'),
    MDollars = ('MDollars','sum')
).reset_index()



HS8_Month_Takh=Exchange_concated_final.groupby(['Year','Month','HS_Type','Tcode',' Takhsis_Type']).agg(
    HS2 = ('HS2','first'),
    HS2_F = ('HS2_F','first'),
    HS2_E = ('HS2_E','first'),
    HS4 = ('HS4','first'),
    HS4_F = ('HS4_F','first'),
    HS4_E = ('HS4_E','first'),
    HS6 = ('HS6','first'),
    HS6_F = ('HS6_F','first'),
    HS6_E = ('HS6_E','first'),
    HS8_F = ('Tdesc','first'),
    MDollars = ('MDollars','sum')
).reset_index()
```
