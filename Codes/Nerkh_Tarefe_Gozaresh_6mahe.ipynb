```python
import pandas as pd
import numpy as np
import os
import openpyxl
import xlrd
```

```python
All_Years = pd.DataFrame()
years = ['1395','1396','1397','1398','1399','1400','1401','1402','1403-6']
for year in years:
    My_Df = pd.read_excel(f'C:/TradeGroup/Reports/Hoghough_Voroudi_Detailed_Timeseries.xlsx',sheet_name=year)

    if year=='1403-6':
        My_Df['Year']=1403
    else:
        My_Df['Year']=int(year)    

    conditions = [
        (My_Df['Hperc'] == 0),             
        ((My_Df['Hperc'] > 0) & (My_Df['Hperc'] <= 5)),
        ((My_Df['Hperc'] > 5) & (My_Df['Hperc'] <= 10)),
        ((My_Df['Hperc'] > 10) & (My_Df['Hperc'] <= 20)),
        ((My_Df['Hperc'] > 20) & (My_Df['Hperc'] <= 50)),
        ((My_Df['Hperc'] > 50) & (My_Df['Hperc'] <= 100)),
        ((My_Df['Hperc'] > 100))
    ]

    choices = [
        'Zero',
        'from 0 to 5',
        'from 5 to 10',
        'from 10 to 20', 
        'from 20 to 50',   
        'from 50 to 100',
        'more than 100'
    ]

    # Use np.select() to generate the new column 'B'
    My_Df['HPerc_Cat'] = np.select(conditions, choices, default='other')

    By_HPerc_Cat = My_Df.groupby(['Year','HPerc_Cat']).agg(
        Value_R = ('Rials','sum'),
        Value_D = ('Dollars','sum'),
        HSum_R = ('Hsum','sum'),
        Codes_Count = ('Tcode','count')
        
        
    ).reset_index()

    All_Years = pd.concat([All_Years,By_HPerc_Cat])

    


All_Years
```
