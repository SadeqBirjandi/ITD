```python
# def replace_arabic_characters(text):
#    text = text.replace('ي', 'ی')
#    text = text.replace('ك', 'ک')


def replace_arabic_characters(df):
    for col in df.columns:
        df[col] = df[col].apply(lambda x: str(x).replace('ي', 'ی'))
        df[col] = df[col].apply(lambda x: str(x).replace('ك', 'ک'))


    return df   
```
