```python
import os
import pandas as pd

# Set the working directory
os.chdir("C:\TradeGroup")
my_wd = os.getcwd()

file_path_Exp_Others = f"{my_wd}\\Data\\Mirror_Data\\Mirror-Others-GExReEx-"
file_path_Imp_Others   = f"{my_wd}\\Data\\Mirror_Data\\Mirror-Others-GImReIm-"

file_path_Exp_Iran = f"{my_wd}\\Data\\Mirror_Data\\Mirror-Iran-GExReEx-"
file_path_Imp_Iran   = f"{my_wd}\\Data\\Mirror_Data\\Mirror-Iran-GImReIm-"


def create_folder(folder_path):
    if not os.path.exists(f'{my_wd}/Data/{folder_path}'):
        os.makedirs(f'{my_wd}/Data/{folder_path}')
    else:
        print(f'Already Exists')

create_folder('Mirror_Data/Results')
```

```python
year = 2021
for year in range(2017,20023):
        
    OExp=pd.read_csv(f"{file_path_Exp_Others}{year}.csv")
    IExp=pd.read_csv(f"{file_path_Exp_Iran}{year}.csv")
    OImp=pd.read_csv(f"{file_path_Imp_Others}{year}.csv")
    IImp=pd.read_csv(f"{file_path_Imp_Iran}{year}.csv")

    OExp=OExp[(OExp['ProductCode'].str.len()==5) | (OExp['ProductCode'].str.len()==6)]
    OExp['ProductCode']= OExp['ProductCode'].astype('int')
    OExp['HS4'] = OExp['ProductCode']//100
    OExp['HS2'] = OExp['ProductCode']//10000
    OExp['HS6'] = OExp['ProductCode']

    OExp = OExp.rename(columns={'ReporterName':'Country'})

    OExp_HS6 = OExp.groupby(['Country','HS6']).agg(
        Year = ('Year','first'),
        HS2 = ('HS2','first'),
        HS4 = ('HS4', 'first'),
        Exp_to_Iran_P = ('TradeValue in 1000 USD','sum')
        ).reset_index()



    OExp_HS4 = OExp.groupby(['Country','HS4']).agg(
        Year = ('Year','first'),
        HS2 = ('HS2','first'),
        Exp_to_Iran_P = ('TradeValue in 1000 USD','sum')

        ).reset_index()




    OExp_HS2 = OExp.groupby(['Country','HS2']).agg(
        Year = ('Year','first'),
        Exp_to_Iran_P = ('TradeValue in 1000 USD','sum')

        ).reset_index()
    



    IExp=IExp[(IExp['ProductCode'].str.len()==5) | (IExp['ProductCode'].str.len()==6)]
    IExp['ProductCode']= IExp['ProductCode'].astype('int')
    IExp['HS4'] = IExp['ProductCode']//100
    IExp['HS2'] = IExp['ProductCode']//10000
    IExp['HS6'] = IExp['ProductCode']

    IExp = IExp.rename(columns={'PartnerName':'Country'})

    IExp_HS6 = IExp.groupby(['Country','HS6']).agg(
        Year = ('Year','first'),
        HS2 = ('HS2','first'),
        HS4 = ('HS4', 'first'),
        Exp_from_Iran_I = ('TradeValue in 1000 USD','sum')
        ).reset_index()



    IExp_HS4 = IExp.groupby(['Country','HS4']).agg(
        Year = ('Year','first'),
        HS2 = ('HS2','first'),
        Exp_from_Iran_I = ('TradeValue in 1000 USD','sum')

        ).reset_index()




    IExp_HS2 = IExp.groupby(['Country','HS2']).agg(
        Year = ('Year','first'),
        Exp_from_Iran_I = ('TradeValue in 1000 USD','sum')

        ).reset_index()   
    








    OImp=OImp[(OImp['ProductCode'].str.len()==5) | (OImp['ProductCode'].str.len()==6)]
    OImp['ProductCode']= OImp['ProductCode'].astype('int')
    OImp['HS4'] = OImp['ProductCode']//100
    OImp['HS2'] = OImp['ProductCode']//10000
    OImp['HS6'] = OImp['ProductCode']

    OImp = OImp.rename(columns={'ReporterName':'Country'})

    OImp_HS6 = OImp.groupby(['Country','HS6']).agg(
        Year = ('Year','first'),
        HS2 = ('HS2','first'),
        HS4 = ('HS4', 'first'),
        Imp_from_Iran_P = ('TradeValue in 1000 USD','sum')
        ).reset_index()



    OImp_HS4 = OImp.groupby(['Country','HS4']).agg(
        Year = ('Year','first'),
        HS2 = ('HS2','first'),
        Imp_from_Iran_P = ('TradeValue in 1000 USD','sum')

        ).reset_index()




    OImp_HS2 = OImp.groupby(['Country','HS2']).agg(
        Year = ('Year','first'),
        Imp_from_Iran_P = ('TradeValue in 1000 USD','sum')

        ).reset_index()
    



    IImp=IImp[(IImp['ProductCode'].str.len()==5) | (IImp['ProductCode'].str.len()==6)]
    IImp['ProductCode']= IImp['ProductCode'].astype('int')
    IImp['HS4'] = IImp['ProductCode']//100
    IImp['HS2'] = IImp['ProductCode']//10000
    IImp['HS6'] = IImp['ProductCode']

    IImp = IImp.rename(columns={'PartnerName':'Country'})

    IImp_HS6 = IImp.groupby(['Country','HS6']).agg(
        Year = ('Year','first'),
        HS2 = ('HS2','first'),
        HS4 = ('HS4', 'first'),
        Imp_to_Iran_I = ('TradeValue in 1000 USD','sum')
        ).reset_index()



    IImp_HS4 = IImp.groupby(['Country','HS4']).agg(
        Year = ('Year','first'),
        HS2 = ('HS2','first'),
        Imp_to_Iran_I = ('TradeValue in 1000 USD','sum')

        ).reset_index()




    IImp_HS2 = IImp.groupby(['Country','HS2']).agg(
        Year = ('Year','first'),
        Imp_to_Iran_I = ('TradeValue in 1000 USD','sum')

        ).reset_index()   



    Iran_Exp_HS6_Mirror = IExp_HS6.merge(OImp_HS6,on=['HS6','HS4','HS2','Year','Country'],how='outer')
    Iran_Exp_HS4_Mirror = IExp_HS4.merge(OImp_HS6,on=['HS4','HS2','Year','Country'],how='outer')
    Iran_Exp_HS2_Mirror = IExp_HS6.merge(OImp_HS2,on=['HS2','Year','Country'],how='outer')



    Iran_Imp_HS6_Mirror = IImp_HS6.merge(OExp_HS6,on=['HS6','HS4','HS2','Year','Country'],how='outer')
    Iran_Imp_HS4_Mirror = IImp_HS4.merge(OExp_HS6,on=['HS4','HS2','Year','Country'],how='outer')
    Iran_Imp_HS2_Mirror = IImp_HS6.merge(OExp_HS2,on=['HS2','Year','Country'],how='outer')



    Iran_Exp_HS6_Mirror = Iran_Exp_HS6_Mirror['Exp_from_Iran_I'].fillna(0)
    Iran_Exp_HS4_Mirror = Iran_Exp_HS4_Mirror['Exp_from_Iran_I'].fillna(0)
    Iran_Exp_HS2_Mirror = Iran_Exp_HS2_Mirror['Exp_from_Iran_I'].fillna(0)


    
    Iran_Exp_HS6_Mirror = Iran_Exp_HS6_Mirror['Exp_from_Iran_I'].fillna(0)
    Iran_Exp_HS4_Mirror = Iran_Exp_HS4_Mirror['Exp_from_Iran_I'].fillna(0)
    Iran_Exp_HS2_Mirror = Iran_Exp_HS2_Mirror['Exp_from_Iran_I'].fillna(0)




    Iran_Imp_HS6_Mirror = Iran_Imp_HS6_Mirror[''].fillna(0)
    Iran_Imp_HS4_Mirror = Iran_Imp_HS4_Mirror.fillna(0)
    Iran_Imp_HS2_Mirror = Iran_Imp_HS2_Mirror.fillna(0)


    Iran_Exp_HS6_Mirror['IranMinusOthers'] = Iran_Exp_HS6_Mirror['Exp_from_Iran_I']- Iran_Exp_HS6_Mirror['Imp_from_Iran_P']
    Iran_Exp_HS4_Mirror['IranMinusOthers'] = Iran_Exp_HS4_Mirror['Exp_from_Iran_I']- Iran_Exp_HS4_Mirror['Imp_from_Iran_P']
    Iran_Exp_HS2_Mirror['IranMinusOthers'] = Iran_Exp_HS2_Mirror['Exp_from_Iran_I']- Iran_Exp_HS2_Mirror['Imp_from_Iran_P']


    Iran_Imp_HS6_Mirror['IranMinusOthers'] = Iran_Imp_HS6_Mirror['Imp_to_Iran_I']- Iran_Imp_HS6_Mirror['Exp_to_Iran_P']
    Iran_Imp_HS4_Mirror['IranMinusOthers'] = Iran_Imp_HS4_Mirror['Imp_to_Iran_I']- Iran_Imp_HS4_Mirror['Exp_to_Iran_P']
    Iran_Imp_HS2_Mirror['IranMinusOthers'] = Iran_Imp_HS2_Mirror['Imp_to_Iran_I']- Iran_Imp_HS2_Mirror['Exp_to_Iran_P']    




    Iran_Exp_HS6_Mirror.to_excel(f"{my_wd}\\Data\\Mirror_Data\\Results\\Iran_Exp_HS6_Mirror_{year}_Shahrivar1403.xlsx",index=False)
    Iran_Exp_HS4_Mirror.to_excel(f"{my_wd}\\Data\\Mirror_Data\\Results\\Iran_Exp_HS4_Mirror_{year}_Shahrivar1403.xlsx",index= False)
    Iran_Exp_HS2_Mirror.to_excel(f"{my_wd}\\Data\\Mirror_Data\\Results\\Iran_Exp_HS2_Mirror_{year}_Shahrivar1403.xlsx",index= False)



    Iran_Imp_HS6_Mirror.to_excel(f"{my_wd}\\Data\\Mirror_Data\\Results\\Iran_Imp_HS6_Mirror_{year}_Shahrivar1403.xlsx",index= False)
    Iran_Imp_HS4_Mirror.to_excel(f"{my_wd}\\Data\\Mirror_Data\\Results\\Iran_Imp_HS4_Mirror_{year}_Shahrivar1403.xlsx",index= False)
    Iran_Imp_HS2_Mirror.to_excel(f"{my_wd}\\Data\\Mirror_Data\\Results\\Iran_Imp_HS2_Mirror_{year}_Shahrivar1403.xlsx",index= False)


    Iran_Exp_HS6_Mirror_New = Iran_Exp_HS6_Mirror[Iran_Exp_HS6_Mirror['Country'] != " World"]
    Iran_Exp_HS4_Mirror_New = Iran_Exp_HS4_Mirror[Iran_Exp_HS4_Mirror['Country'] != " World"]
    Iran_Exp_HS2_Mirror_New = Iran_Exp_HS2_Mirror[Iran_Exp_HS2_Mirror['Country'] != " World"]


    Iran_Imp_HS6_Mirror_New = Iran_Imp_HS6_Mirror[Iran_Imp_HS6_Mirror['Country'] != " World"]
    Iran_Imp_HS4_Mirror_New = Iran_Imp_HS4_Mirror[Iran_Imp_HS4_Mirror['Country'] != " World"]
    Iran_Imp_HS2_Mirror_New = Iran_Imp_HS2_Mirror[Iran_Imp_HS2_Mirror['Country'] != " World"]



    Iran_Exp_HS6_Mirror_Countryless = Iran_Exp_HS6_Mirror_New.groupby(['HS2','HS4','HS6']).agg(
        Exp_from_Iran_I = ('Exp_from_Iran_I', 'sum'),
        Imp_from_Iran_P = ('Imp_from_Iran_P', 'sum'),
        IranMinusOthers = ('IranMinusOthers','sum')
        ).reset_index()
    

    
    Iran_Exp_HS4_Mirror_Countryless = Iran_Exp_HS4_Mirror_New.groupby(['HS2','HS4']).agg(
        Exp_from_Iran_I = ('Exp_from_Iran_I', 'sum'),
        Imp_from_Iran_P = ('Imp_from_Iran_P', 'sum'),
        IranMinusOthers = ('IranMinusOthers','sum')
        ).reset_index()
    


    
    Iran_Exp_HS2_Mirror_Countryless = Iran_Exp_HS2_Mirror_New.groupby(['HS2']).agg(
        Exp_from_Iran_I = ('Exp_from_Iran_I', 'sum'),
        Imp_from_Iran_P = ('Imp_from_Iran_P', 'sum'),
        IranMinusOthers = ('IranMinusOthers','sum')
        ).reset_index()
    



    Iran_Imp_HS6_Mirror_Countryless = Iran_Imp_HS6_Mirror_New.groupby(['HS2','HS4','HS6']).agg(
        Imp_to_Iran_I = ('Imp_to_Iran_I', 'sum'),
        Exp_to_Iran_P = ('Exp_to_Iran_P', 'sum'),
        IranMinusOthers = ('IranMinusOthers','sum')
        ).reset_index()
    

    
    Iran_Imp_HS4_Mirror_Countryless = Iran_Imp_HS4_Mirror_New.groupby(['HS2','HS4']).agg(
        Imp_to_Iran_I = ('Imp_to_Iran_I', 'sum'),
        Exp_to_Iran_P = ('Exp_to_Iran_P', 'sum'),
        IranMinusOthers = ('IranMinusOthers','sum')
        ).reset_index()
    


    
    Iran_Imp_HS2_Mirror_Countryless = Iran_Imp_HS2_Mirror_New.groupby(['HS2']).agg(
        Imp_to_Iran_I = ('Imp_to_Iran_I', 'sum'),
        Exp_to_Iran_P = ('Exp_to_Iran_P', 'sum'),
        IranMinusOthers = ('IranMinusOthers','sum')
        ).reset_index()



    Iran_Exp_HS6_Mirror_Countryless.to_excel(f"{my_wd}\\Data\\Mirror_Data\\Results\\Iran_Exp_HS6_Mirror_{year}_Countryless_Shahrivar1403.xlsx",index=False)
    Iran_Exp_HS4_Mirror_Countryless.to_excel(f"{my_wd}\\Data\\Mirror_Data\\Results\\Iran_Exp_HS4_Mirror_{year}_Countryless_Shahrivar1403.xlsx",index= False)
    Iran_Exp_HS2_Mirror_Countryless.to_excel(f"{my_wd}\\Data\\Mirror_Data\\Results\\Iran_Exp_HS2_Mirror_{year}_Countryless_Shahrivar1403.xlsx",index= False)



    Iran_Imp_HS6_Mirror_Countryless.to_excel(f"{my_wd}\\Data\\Mirror_Data\\Results\\Iran_Imp_HS6_Mirror_{year}_Countryless_Shahrivar1403.xlsx",index= False)
    Iran_Imp_HS4_Mirror_Countryless.to_excel(f"{my_wd}\\Data\\Mirror_Data\\Results\\Iran_Imp_HS4_Mirror_{year}_Countryless_Shahrivar1403.xlsx",index= False)
    Iran_Imp_HS2_Mirror_Countryless.to_excel(f"{my_wd}\\Data\\Mirror_Data\\Results\\Iran_Imp_HS2_Mirror_{year}_Countryless_Shahrivar1403.xlsx",index= False)

    


```
