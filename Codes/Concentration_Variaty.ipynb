```python
import pandas as pd
import numpy as np
import os
```

```python
os.chdir("C:\TradeGroup")
my_wd = os.getcwd()
new_directory = my_wd

def create_folder(folder_path):
    if not os.path.exists(f'{new_directory}/{folder_path}'):
        os.makedirs(f'{new_directory}/{folder_path}')
    else:
        print(f'Already Exists')

create_folder('Data/Indices')
```

```python
year = 1383
for year in range(1383,1403):
    exports_path = os.path.join(my_wd, f"Data/Cleaned_Data/Y{year}Exports.xlsx")
    imports_path = os.path.join(my_wd, f"Data/Cleaned_Data/Y{year}Imports.xlsx")

    print(year)
    Exports = pd.read_excel(exports_path)
    Exports.dropna(axis=0, subset= ["Country"], inplace=True)

    #Imports = pd.read_excel(imports_path)
    Exports['Tcode']=Exports['Tcode'].astype(int)
    Exports['HS6']=(Exports['Tcode']//100).astype(int)

    HSs_agg=Exports.groupby("Tcode").agg(
        Dollars = ("Dollars","sum")
    ).reset_index()

    HSs_agg_HS6=Exports.groupby("HS6").agg(
        Dollars = ("Dollars","sum")
    ).reset_index()    

    HSs_agg['Share2'] = (HSs_agg['Dollars']/HSs_agg['Dollars'].sum())**2 
    HSs_agg_HS6['Share2'] = (HSs_agg_HS6['Dollars']/HSs_agg_HS6['Dollars'].sum())**2 


    HSs_agg_HS6_FIRST20 = HSs_agg_HS6.sort_values(by='Share2', ascending=False)

# Keep only the first 20 rows
    HSs_agg_HS6_FIRST20 = HSs_agg_HS6_FIRST20.head(20)

    HSs_agg_HS6_FIRST20.to_excel(f'{my_wd}\Data\Indices\Concentrationfirst20Y{year}.xlsx',index=False)

    HS6_unique_values = len(Exports["Tcode"].unique())
    HSs_agg['Under10']= np.where(HSs_agg['Dollars']<10000,  1, 0)
    HSs_agg['f10to100'] = np.where((HSs_agg['Dollars']>=10000) & (HSs_agg['Dollars']<100000) , 1, 0)
    HSs_agg['Over10'] = np.where(HSs_agg['Dollars']>=10000, 1, 0)
    HSs_agg['Over100'] = np.where(HSs_agg['Dollars']>=100000, 1, 0)



    HSs_agg_HS6['Under10']= np.where(HSs_agg_HS6['Dollars']<10000,  1, 0)
    HSs_agg_HS6['f10to100'] = np.where((HSs_agg_HS6['Dollars']>=10000) & (HSs_agg_HS6['Dollars']<100000) , 1, 0)
    HSs_agg_HS6['Over10'] = np.where(HSs_agg_HS6['Dollars']>=10000, 1, 0)
    HSs_agg_HS6['Over100'] = np.where(HSs_agg_HS6['Dollars']>=100000, 1, 0)

    if year == 1383:
        HS6_uniques_series = pd.DataFrame()
        HS6_uniques_series_HS6 = pd.DataFrame()
        
    HS6_uniques = pd.DataFrame({"Year": [year], "Count": [len(Exports["Tcode"].unique())] , "veriaty": sum(HSs_agg['Share2']), "veriaty_HS6": sum(HSs_agg_HS6['Share2']),
                                "Under10": sum(HSs_agg['Under10']), "f10to100": sum(HSs_agg['f10to100']) , "Over10": sum(HSs_agg['Over10']), "Over100": sum(HSs_agg['Over100']) })
    
    HS6_uniques_HS6 = pd.DataFrame({"Year": [year], "Count": [len(HSs_agg_HS6["HS6"].unique())] , "veriaty": sum(HSs_agg_HS6['Share2']), "veriaty_HS6": sum(HSs_agg_HS6['Share2']),
                                "Under10": sum(HSs_agg_HS6['Under10']), "f10to100": sum(HSs_agg_HS6['f10to100']) , "Over10": sum(HSs_agg_HS6['Over10']), "Over100": sum(HSs_agg_HS6['Over100']) })

    HS6_uniques_series = pd.concat([HS6_uniques_series,HS6_uniques])

    HS6_uniques_series_HS6 = pd.concat([HS6_uniques_series_HS6,HS6_uniques_HS6])


```
