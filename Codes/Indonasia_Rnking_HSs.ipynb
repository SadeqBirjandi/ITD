```python
import pandas as pd
import numpy as np


Iran_to_Indonasia=pd.read_excel(f"C:/TradeGroup/Reports/Iran-Indonasia-Be.xlsx")
Iran_to_Indonasia['شماره تعرفه'].fillna(0,inplace=True)
Iran_to_Indonasia['HS8']=Iran_to_Indonasia['شماره تعرفه'].astype(int)
Iran_Exports_Codes = pd.unique(Iran_to_Indonasia['HS8'])
Iran_Exports_Codes

My_Excel_File=pd.ExcelFile(f"C:/TradeGroup/Reports/Iran-Indonasia-Alef.xlsx")
dfs = []
for sheet_name in My_Excel_File.sheet_names:
    df= My_Excel_File.parse(sheet_name)
    dfs.append(df)

The_Commodities = pd.concat(dfs,ignore_index=True)
The_Commodities.rename(columns={'شماره تعرفه':"HS8"},inplace=True)
The_Commodities['HS8']=pd.to_numeric(The_Commodities['HS8'])
Iran_Imports_Codes = pd.unique(The_Commodities['HS8'])
#Iran_Imports_Codes


```

```python
Iran_Exports_Codes = Iran_Exports_Codes//100
```

```python
year=1398
baseyear=1398
for year in range(1398,1403):
    Exports = pd.read_excel(f'C:/TradeGroup/Data/Cleaned_Data/Y{year}Exports.xlsx')
    Exports_Indonasia = Exports[Exports['Country']=="اندونزي"]

    Exports_Indonasia['Tcode'] = Exports_Indonasia['Tcode']//100

    Exports_Indonasia['Tarjihi']=np.where(np.isin(Exports_Indonasia['Tcode'], Iran_Exports_Codes),1,0 )


    Exports_Idn=Exports_Indonasia.groupby('Tcode').agg(
                Tarjihi= ('Tarjihi','first'),
    #             Tdesc = ('Tdesc','first'),
                Dollars_Tot=('Dollars','sum')
                
                
                #   Tcode= ('Tcode','first')
            ).reset_index()

    Exports_Idn['Share']=Exports_Idn['Dollars_Tot']/Exports_Idn['Dollars_Tot'].sum()
    #Exports_Idn.columns = Exports_Idn.columns.str.replace(['Dollars_Tot','Share'],[f'Dollars_{year}',f'Share_{year}'])

    Exports_Idn.columns = Exports_Idn.columns.str.replace('Dollars_Tot', f'Dollars_{year}')
    Exports_Idn.columns = Exports_Idn.columns.str.replace('Share', f'Share_{year}')

    Exports_Idn[f'rank_{year}'] = Exports_Idn[f'Dollars_{year}'].rank(ascending= False)

    if year == baseyear:
        Exports_Idn_Series=Exports_Idn
    else:
        Exports_Idn_Series = Exports_Idn_Series.merge(Exports_Idn, on  = ['Tcode', 'Tarjihi'], how="outer")
        Exports_Idn_Series.fillna(0, inplace = True)    

Exports_Idn_Series
```

```python
column_names = [f'Dollars_{year}' for year in range(1398,1403)]

Exports_Idn_Series[f'Total_Dollars']=Exports_Idn_Series[column_names].sum(axis=1)
Exports_Idn_Series[f'Total_Share'] = Exports_Idn_Series['Total_Dollars']/Exports_Idn_Series['Total_Dollars'].sum()

Exports_Idn_Series['Rank']=Exports_Idn_Series['Total_Dollars'].rank(ascending= False)

#Exports_Idn_Series.to_excel(f"C:/TradeGroup/Reports/Ian_Indonasia_Share_Ranking.xlsx", index= False)
```
