```python
PistachiosShelled= pd.read_excel(f'{my_wd}/Data/Indices/PistachiosShelled_Importers.xlsx')
PistachiosInShell= pd.read_excel(f'{my_wd}/Data/Indices/PistachiosInShell_Importers.xlsx')
PistachiosInShellShelled= pd.read_excel(f'{my_wd}/Data/Indices/PistachiosInShellShelled_Importers.xlsx')




Iran_PistachiosShelled= pd.read_excel(f'{my_wd}/Data/Indices/Iran_PistachiosShelled_Exports.xlsx')
Iran_PistachiosInShell= pd.read_excel(f'{my_wd}/Data/Indices/Iran_PistachiosInShell_Exports.xlsx')
Iran_PistachiosInShellShelled= pd.read_excel(f'{my_wd}/Data/Indices/Iran_PistachiosInShellShelled_Exports.xlsx')





PSC=PistachiosShelled.columns
PISC=PistachiosInShell.columns
PISSC=PistachiosInShellShelled.columns



Iran_PSC = Iran_PistachiosShelled.columns
Iran_PISC = Iran_PistachiosInShell.columns
Iran_PISSC = Iran_PistachiosInShellShelled.columns


Importers_PS = set(PistachiosShelled['Importers'])
Importers_PIS = set(PistachiosInShell['Importers'])
Importers_PISS = set(PistachiosInShellShelled['Importers'])



Iran_PS = set(Iran_PistachiosShelled['Importers'])
Iran_PIS = set(Iran_PistachiosInShell['Importers'])
Iran_PISS = set(Iran_PistachiosInShellShelled['Importers'])


Importers = Importers_PS.union(Importers_PIS)
Importers = Importers.union(Importers_PISS)
Importers = list(Importers)

Iran_Importers = Iran_PS.union(Iran_PIS)
Iran_Importers = Iran_Importers.union(Iran_PISS)
Iran_Importers = list(Iran_Importers)

Pistachios_Importers=pd.DataFrame({"Importers":Importers})

Iran_Pistachio_Importers=pd.DataFrame({"Importers":Iran_Importers})


Imports_merged = PistachiosShelled.merge(PistachiosInShell,on="Importers",how="outer",suffixes=(' PS',' PIS'))
Imports_merged = Imports_merged.merge(PistachiosInShellShelled,on="Importers",how="outer")
Imports_merged.fillna(0,inplace=True)


Exports_merged = Iran_PistachiosShelled.merge(Iran_PistachiosInShell,on="Importers",how="outer",suffixes=(' PS',' PIS'))
Exports_merged = Exports_merged.merge(Iran_PistachiosInShellShelled,on="Importers",how="outer")
Exports_merged.fillna(0,inplace=True)

for col in PSC[1:]:
    Imports_merged[col]=Imports_merged[f'{col} PS']+Imports_merged[f'{col} PIS']+Imports_merged[f'{col}']

for col in Iran_PSC[1:]:
    Exports_merged[col]=Exports_merged[f'{col} PS']+Exports_merged[f'{col} PIS']+Exports_merged[f'{col}']

Pistachios_Total=Imports_merged[PSC]

Iran_Pistachios_Total = Exports_merged[Iran_PSC]


Pistachios_Total.to_excel(f'{my_wd}/Data/Indices/Pistachios_Importers.xlsx',index=False)

Iran_Pistachios_Total.to_excel(f'{my_wd}/Data/Indices/Iran_Pistachios_Exports.xlsx',index=False)

```

```python
#Commodities = ['Date','Saffron','Carpet']
Com_New = ['PistachiosInShell','PistachiosShelled','PistachiosInShellShelled','Pistachios']
for com in Com_New:
    Com_World_Importers= pd.read_excel(f'{my_wd}/Data/Indices/{com}_Importers.xlsx')


    Com_World_Importers.columns=Com_World_Importers.columns.str.replace("Imported value in ","")
    Com_World_Importers.replace(0, np.nan, inplace=True)
    Com_Total_Importers=pd.DataFrame(Com_World_Importers.count(),columns=[f"World_{com}"]).reset_index()
    Com_Total_Importers=Com_Total_Importers.iloc[1:]

    #Com_Total_Importers.rename(columns={Com_Total_Importers[0]: "Year"})

    Com_Total_Importers.rename(columns={Com_Total_Importers.columns[0]:"Year"},inplace=True)

    Com_Iran_Importers= pd.read_excel(f'{my_wd}/Data/Indices/Iran_{com}_Exports.xlsx')

    Com_Iran_Importers.columns=Com_Iran_Importers.columns.str.replace("Exported value in ","")
    Com_Iran_Importers.replace(0, np.nan, inplace=True)
    Com_Iran_Importers_Exp=pd.DataFrame(Com_Iran_Importers.count(),columns=[f"Iran_{com}"]).reset_index()
    Com_Iran_Importers_Exp=Com_Iran_Importers_Exp.iloc[1:]

    #Com_Iran_Importers_Exp.rename(columns={Com_Iran_Importers_Exp[0]: "Year"})
    Com_Iran_Importers_Exp.rename(columns={Com_Iran_Importers_Exp.columns[0]:"Year"},inplace=True)

    Com_Total_Importers = Com_Total_Importers.merge(Com_Iran_Importers_Exp,on=['Year'], how= 'outer')
    Com_Total_Importers[f'{com}_to_World']=Com_Total_Importers[f'Iran_{com}']/Com_Total_Importers[f'World_{com}']

    

    if com == Com_New[0]:
        Coms_Importers = Com_Total_Importers
    else:
        Coms_Importers = Coms_Importers.merge(Com_Total_Importers, on=['Year'])    


Coms_Importers_transposed=Coms_Importers.transpose().reset_index()
Coms_Importers_transposed.columns=Coms_Importers_transposed.iloc[0]
Coms_Importers_transposed=Coms_Importers_transposed[1:]
Coms_Importers_transposed



Coms_Importers_transposed.to_excel(f'{my_wd}/Data/Indices/Commudities_Peneteration.xlsx')
       
```
