```python
import os
import pandas as pd

# Set the working directory
os.chdir("C:\TradeGroup")
my_wd = os.getcwd()

file_path_Exp_Others = f"{my_wd}\\Data\\Mirror_Data\\Mirror-Others-GExReEx-"
file_path_Imp_Others   = f"{my_wd}\\Data\\Mirror_Data\\Mirror-Others-GImReIm-"

file_path_Exp_Iran = f"{my_wd}\\Data\\Mirror_Data\\Mirror-Iran-GExReEx-"
file_path_Imp_Iran   = f"{my_wd}\\Data\\Mirror_Data\\Mirror-Iran-GImReIm-"


def create_folder(folder_path):
    if not os.path.exists(f'{my_wd}/Data/{folder_path}'):
        os.makedirs(f'{my_wd}/Data/{folder_path}')
    else:
        print(f'Already Exists')

create_folder('Mirror_Data/Results')
```

```python
for year in range(2017,2022):
    OExp=pd.read_csv(f"{file_path_Exp_Others}{year}.csv")
    IExp=pd.read_csv(f"{file_path_Exp_Iran}{year}.csv")
    OImp=pd.read_csv(f"{file_path_Imp_Others}{year}.csv")
    IImp=pd.read_csv(f"{file_path_Imp_Iran}{year}.csv")

    OExp=OExp[(OExp['ProductCode'].str.len()==5) | (OExp['ProductCode'].str.len()==6)]
    OExp['ProductCode']= OExp['ProductCode'].astype('int')

    OExp['HS6'] = OExp['ProductCode']

    OExp = OExp.rename(columns={'ReporterName':'Country','TradeValue in 1000 USD':'Dollars_O'})



    IImp=IImp[(IImp['ProductCode'].str.len()==5) | (IImp['ProductCode'].str.len()==6)]
    IImp['ProductCode']= IImp['ProductCode'].astype('int')

    IImp['HS6'] = IImp['ProductCode']

    IImp = IImp.rename(columns={'PartnerName':'Country','TradeValue in 1000 USD':'Dollars_I'})


    OExp = OExp[['Country','HS6','Dollars_O']]
    IImp = IImp[['Country','HS6','Dollars_I']]


    Iran_Imp = OExp.merge(IImp,on=["Country",'HS6'],how='outer')
    Iran_Imp['Year'] = year
    Iran_Imp = Iran_Imp.fillna(0)

    Iran_Imp['IranMinusOthers'] = Iran_Imp['Dollars_I']-Iran_Imp['Dollars_O']


    Iran_Imp = Iran_Imp[Iran_Imp['Country']!=" World"]




    IExp=IExp[(IExp['ProductCode'].str.len()==5) | (IExp['ProductCode'].str.len()==6)]
    IExp['ProductCode']= IExp['ProductCode'].astype('int')

    IExp['HS6'] = IExp['ProductCode']

    IExp = IExp.rename(columns={'PartnerName':'Country','TradeValue in 1000 USD':'Dollars_I'})



    OImp=OImp[(OImp['ProductCode'].str.len()==5) | (OImp['ProductCode'].str.len()==6)]
    OImp['ProductCode']= OImp['ProductCode'].astype('int')

    OImp['HS6'] = OImp['ProductCode']

    OImp = OImp.rename(columns={'ReporterName':'Country','TradeValue in 1000 USD':'Dollars_O'})


    IExp = IExp[['Country','HS6','Dollars_I']]
    OImp = OImp[['Country','HS6','Dollars_O']]


    Iran_Exp = IExp.merge(OImp,on=["Country",'HS6'],how='outer')
    Iran_Exp['Year'] = year
    Iran_Exp = Iran_Exp.fillna(0)

    Iran_Exp['IranMinusOthers'] = Iran_Exp['Dollars_I']-Iran_Exp['Dollars_O']

    Iran_Exp = Iran_Exp[Iran_Exp['Country']!=" World"]



    Iran_Exp['HS4']=Iran_Exp['HS6']//100
    Iran_Exp['HS2']= Iran_Exp['HS6']//10000


    Iran_Imp['HS4']=Iran_Imp['HS6']//100
    Iran_Imp['HS2']= Iran_Imp['HS6']//10000


    Iran_Exp.to_excel(f'{my_wd}\\Data\\Mirror_Data\\Results\\Iran_Exp_HS6_Mirror_{year}_Complete_Shahrivar1403.xlsx',index=False)

    Iran_Imp.to_excel(f'{my_wd}\\Data\\Mirror_Data\\Results\\Iran_Imp_HS6_Mirror_{year}_Complete_Shahrivar1403.xlsx',index = False)


    Iran_Exp_Country = Iran_Exp.groupby('Country').agg(
        Dollars_I = ('Dollars_I','sum'),
        Dollars_O = ('Dollars_O', 'sum'),
        IranMinusOthers = ('IranMinusOthers','sum')

     ).reset_index()



    Iran_Imp_Country = Iran_Imp.groupby('Country').agg(
        Dollars_I = ('Dollars_I','sum'),
        Dollars_O = ('Dollars_O', 'sum'),
        IranMinusOthers = ('IranMinusOthers','sum')

     ).reset_index()



    Iran_Exp_HS6 = Iran_Exp.groupby('HS6').agg(
        Dollars_I = ('Dollars_I','sum'),
        Dollars_O = ('Dollars_O', 'sum'),
        IranMinusOthers = ('IranMinusOthers','sum')

     ).reset_index()



    Iran_Imp_HS6 = Iran_Imp.groupby('HS6').agg(
        Dollars_I = ('Dollars_I','sum'),
        Dollars_O = ('Dollars_O', 'sum'),
        IranMinusOthers = ('IranMinusOthers','sum')

     ).reset_index()




    Iran_Exp_HS4 = Iran_Exp.groupby('HS4').agg(
        Dollars_I = ('Dollars_I','sum'),
        Dollars_O = ('Dollars_O', 'sum'),
        IranMinusOthers = ('IranMinusOthers','sum')

     ).reset_index()



    Iran_Imp_HS4 = Iran_Imp.groupby('HS4').agg(
        Dollars_I = ('Dollars_I','sum'),
        Dollars_O = ('Dollars_O', 'sum'),
        IranMinusOthers = ('IranMinusOthers','sum')

     ).reset_index()




    Iran_Exp_HS2 = Iran_Exp.groupby('HS2').agg(
        Dollars_I = ('Dollars_I','sum'),
        Dollars_O = ('Dollars_O', 'sum'),
        IranMinusOthers = ('IranMinusOthers','sum')

     ).reset_index()



    Iran_Imp_HS2 = Iran_Imp.groupby('HS2').agg(
        Dollars_I = ('Dollars_I','sum'),
        Dollars_O = ('Dollars_O', 'sum'),
        IranMinusOthers = ('IranMinusOthers','sum')

     ).reset_index()





    Iran_Exp_Agg = Iran_Exp.groupby('Year').agg(
        Dollars_I = ('Dollars_I','sum'),
        Dollars_O = ('Dollars_O', 'sum'),
        IranMinusOthers = ('IranMinusOthers','sum')

     ).reset_index()



    Iran_Imp_Agg = Iran_Imp.groupby('Year').agg(
        Dollars_I = ('Dollars_I','sum'),
        Dollars_O = ('Dollars_O', 'sum'),
        IranMinusOthers = ('IranMinusOthers','sum')

     ).reset_index()



    Iran_Exp_Country.to_excel(f'{my_wd}\\Data\\Mirror_Data\\Results\\Iran_Exp_Country_Mirror_{year}_Complete_Shahrivar1403.xlsx',index=False)

    Iran_Imp_Country.to_excel(f'{my_wd}\\Data\\Mirror_Data\\Results\\Iran_Imp_Country_Mirror_{year}_Complete_Shahrivar1403.xlsx',index = False)


    Iran_Exp_HS2.to_excel(f'{my_wd}\\Data\\Mirror_Data\\Results\\Iran_Exp_HS2_Mirror_{year}_Complete_Shahrivar1403.xlsx',index=False)

    Iran_Imp_HS2.to_excel(f'{my_wd}\\Data\\Mirror_Data\\Results\\Iran_Imp_HS2_Mirror_{year}_Complete_Shahrivar1403.xlsx',index = False)


    Iran_Exp_HS4.to_excel(f'{my_wd}\\Data\\Mirror_Data\\Results\\Iran_Exp_HS4_Mirror_{year}_Complete_Shahrivar1403.xlsx',index=False)

    Iran_Imp_HS4.to_excel(f'{my_wd}\\Data\\Mirror_Data\\Results\\Iran_Imp_HS4_Mirror_{year}_Complete_Shahrivar1403.xlsx',index = False)


    Iran_Exp_HS6.to_excel(f'{my_wd}\\Data\\Mirror_Data\\Results\\Iran_Exp_HS6_Mirror_{year}_Complete_Shahrivar1403.xlsx',index=False)

    Iran_Imp_HS6.to_excel(f'{my_wd}\\Data\\Mirror_Data\\Results\\Iran_Imp_HS6_Mirror_{year}_Complete_Shahrivar1403.xlsx',index = False)

    if year ==2017:
        Iran_Exp_Agg_Series = Iran_Exp_Agg
        Iran_Imp_Agg_Series = Iran_Imp_Agg

    else:
        Iran_Exp_Agg_Series = pd.concat([Iran_Exp_Agg_Series,Iran_Exp_Agg]) 
        Iran_Imp_Agg_Series = pd.concat([Iran_Imp_Agg_Series,Iran_Imp_Agg]) 


Iran_Exp_Agg_Series.to_excel(f'{my_wd}\\Data\\Mirror_Data\\Results\\Iran_Exp_Mirror_Complete_Shahrivar1403.xlsx',index=False)

Iran_Imp_Agg_Series.to_excel(f'{my_wd}\\Data\\Mirror_Data\\Results\\Iran_Imp_Mirror_Complete_Shahrivar1403.xlsx',index = False)



    

    
```
