```python
import pandas as pd
import numpy as np
import os
```

```python
# Set the working directory
os.chdir("C:\TradeGroup")
my_wd = os.getcwd()
new_directory = my_wd

def create_folder(folder_path):
    if not os.path.exists(f'{new_directory}/{folder_path}'):
        os.makedirs(f'{new_directory}/{folder_path}')
    else:
        print(f'Already Exists')

create_folder('Data/Indices')
```

```python
baseyear=1395
for year in range(1395, 1403):
    # Clear existing variables
    try:
        del Exports
    except NameError:
        pass

    try:
        del Imports
    except NameError:
        pass

    # Load data
    exports_path = os.path.join(my_wd, f"Data/Cleaned_Data/Y{year}Exports.xlsx")
    imports_path = os.path.join(my_wd, f"Data/Cleaned_Data/Y{year}Imports.xlsx")

    print(year)

    Exports = pd.read_excel(exports_path)
    Imports = pd.read_excel(imports_path)

    Imports.dropna(subset=['Tcode'],inplace=True)
    Exports.dropna(subset=['Tcode'],inplace=True)

   
    Imports['Dollars']=Imports['Dollars'].astype(float)
    Imports['Weight']=Imports['Weight'].astype(float)

    Exports['Dollars']=Exports['Dollars'].astype(float)
    Exports['Weight']=Exports['Weight'].astype(float)    

    
    Imports['Weight_DPrice']=Imports['Dollars']/Imports['Weight']
    Exports['Weight_DPrice']=Exports['Dollars']/Exports['Weight']

    Imports['Weight_RPrice']=Imports['Rials']/Imports['Weight']
    Exports['Weight_RPrice']=Exports['Rials']/Exports['Weight']

    Imports = Imports[Imports['Weight']!=0]    
    Exports = Exports[Exports['Weight']!=0]

    
    #Weight_Exports_Custom = Exports.groupby("Custom").agg(
    #        Weight_DPrice=("Weight_DPrice", "mean"),
    #        Weight_RPrice=("Weight_RPrice", "mean")
    #        
            #,
            #Rials_Exp=("Rials", "sum"),
            #Dollars_Exp=("Dollars", "sum")
    #    ).reset_index()
    

    Weight_Exports_Country = Exports.groupby("Country").apply(
             lambda x: pd.Series({
                 'Weight_DPrice':
             np.average(x['Weight_DPrice'],
                        weights=x['Weight']),
                'Weight_RPrice':
            np.average(x['Weight_RPrice'],
                       weights=x['Weight']), 
                'Rials_Exp':
            np.sum(x["Rials"]),
                'Dollars_Exp':
            np.sum(x['Dollars'])          
            })
        ).reset_index()
    print(f'Ex_Country_{year}_Done')




   # Weight_Exports_TCode = Exports.groupby(["Tcode",'Tdesc']).agg(
   #         Weight_DPrice=("Weight_DPrice", "mean"),
   #         Weight_RPrice=("Weight_RPrice", "mean")
            #,
            #Rials_Exp=("Rials", "sum"),
            #Dollars_Exp=("Dollars", "sum")
   #     ).reset_index()
   # 
    Weight_Exports_TCode = Exports.groupby(["Tcode",'Tdesc']).apply(
            lambda x: pd.Series({
                'Weight_DPrice':
            np.average(x['Weight_DPrice'],
                       weights=x['Weight']),
                'Weight_RPrice':
            np.average(x['Weight_RPrice'],
                       weights=x['Weight'])         
            })
        ).reset_index()
    print(f'Ex_TCode_{year}_Done')   




    #Weight_Imports_Custom = Imports.groupby("Custom").agg(
    #        Weight_DPrice=("Weight_DPrice", "mean"),
    #        Weight_RPrice=("Weight_RPrice", "mean")
            #,
            #Rials_Exp=("Rials", "sum"),
            #Dollars_Exp=("Dollars", "sum")
    #    ).reset_index()
    

    # Weight_Imports_Country = Imports.groupby("Country").agg(
    #        Weight_DPrice=("Weight_DPrice", "mean"),
    #        Weight_RPrice=("Weight_RPrice", "mean")
    #         ,
    #         Rials_Exp=("Rials", "sum"),
    #         Dollars_Exp=("Dollars", "sum")
    #    ).reset_index().assign(Year=year)


    Weight_Imports_Country = Imports.groupby("Country").apply(
            lambda x: pd.Series({
                'Weight_DPrice':
            np.average(x['Weight_DPrice'],
                       weights=x['Weight']),
                'Weight_RPrice':
            np.average(x['Weight_RPrice'],
                       weights=x['Weight']), 
                'Rials_Imp':
            np.sum(x["Rials"]),
                'Dollars_Imp':
                np.sum(x['Dollars'])          
            })
        ).reset_index()
    print(f'Im_Country_{year}_Done')

   # Weight_Imports_TCode = Imports.groupby(["Tcode",'Tdesc']).agg(
   #         Weight_DPrice=("Weight_DPrice", "mean"),
   #         Weight_RPrice=("Weight_RPrice", "mean")
            #,
            #Rials_Exp=("Rials", "sum"),
            #Dollars_Exp=("Dollars", "sum")
   #     ).reset_index().assign(Year=year)
    Weight_Imports_TCode = Imports.groupby(["Tcode",'Tdesc']).apply(
            lambda x: pd.Series({
                'Weight_DPrice':
            np.average(x['Weight_DPrice'],
                       weights=x['Weight']),
                'Weight_RPrice':
            np.average(x['Weight_RPrice'],
                       weights=x['Weight'])         
            })
        ).reset_index().assign(Year=year)
    print(f'Ex_TCode_{year}_Done') 
    

#    Weight_Exports_Country_Tcode = Exports.groupby(["Country","Tcode",'Tdesc']).agg(
#            Weight_DPrice=("Weight_DPrice", "mean"),
#            Weight_RPrice=("Weight_RPrice", "mean")
#            ).reset_index().assign(Year=year)
    
    Weight_Exports_Country_Tcode = Exports.groupby(["Country","Tcode",'Tdesc']).apply(
            lambda x: pd.Series({
                'Weight_DPrice':
            np.average(x['Weight_DPrice'],
                       weights=x['Weight']),
                'Weight_RPrice':
            np.average(x['Weight_RPrice'],
                       weights=x['Weight'])    
            })
            ).reset_index().assign(Year=year)
    print(f'Ex_Country_Tcode_{year}_Done')



#    Weight_Imports_Country_Tcode = Imports.groupby(["Country","Tcode",'Tdesc']).agg(
#            Weight_DPrice=("Weight_DPrice", "mean"),
#            Weight_RPrice=("Weight_RPrice", "mean")
#            ).reset_index().assign(Year=year) 



    Weight_Imports_Country_Tcode = Imports.groupby(["Country","Tcode",'Tdesc']).apply(
            lambda x: pd.Series({
                'Weight_DPrice':
            np.average(x['Weight_DPrice'],
                       weights=x['Weight']),
                'Weight_RPrice':
            np.average(x['Weight_RPrice'],
                       weights=x['Weight'])    
            })
            ).reset_index().assign(Year=year)
    print(f'Im_Country_Tcode_{year}_Done') 
    
  #  Weight_Exports_TCode['Year']=Weight_Exports_Country['Year']=Weight_Exports_Custom['Year']= \
   # Weight_Imports_Country['Year']=Weight_Imports_Custom['Year']=Weight_Imports_TCode['Year']= \
    #Weight_Exports_Country_Tcode['Year']=Weight_Imports_Country_Tcode['Year']=year
    
    if year==baseyear:
        Imports_Country_Rprice=Weight_Imports_Country.loc[:,['Country','Weight_RPrice']].rename(columns={'Weight_RPrice':"RPrice_Imp_"+str(year)}) 
        Exports_Country_Rprice=Weight_Exports_Country.loc[:,['Country','Weight_RPrice']].rename(columns={'Weight_RPrice':"RPrice_Exp_"+str(year)})


    #    Imports_Custom_Rprice=Weight_Imports_Custom.loc[:,['Custom','Weight_RPrice']].rename(columns={'Weight_RPrice':"RPrice_Imp_"+str(year)})
    #    Exports_Custom_Rprice=Weight_Exports_Custom.loc[:,['Custom','Weight_RPrice']].rename(columns={'Weight_RPrice':"RPrice_Exp_"+str(year)})


        Imports_TCode_Rprice=Weight_Imports_TCode.loc[:,['Tcode','Tdesc','Weight_RPrice']].rename(columns={'Weight_RPrice':"RPrice_Imp_"+str(year)})
        Exports_TCode_Rprice=Weight_Exports_TCode.loc[:,['Tcode','Tdesc','Weight_RPrice']].rename(columns={'Weight_RPrice':"RPrice_Exp_"+str(year)})


        Imports_Country_TCode_RPrice=Weight_Imports_Country_Tcode.loc[:,['Country','Tcode','Tdesc','Weight_RPrice']].rename(columns={'Weight_RPrice':"RPrice_Imp_"+str(year)})
        Exports_Country_TCode_RPrice=Weight_Exports_Country_Tcode.loc[:,['Country','Tcode','Tdesc','Weight_RPrice']].rename(columns={'Weight_RPrice':"RPrice_Exp_"+str(year)})   


        Imports_Country_Dprice=Weight_Imports_Country.loc[:,['Country','Weight_DPrice']].rename(columns={'Weight_DPrice':"DPrice_Imp_"+str(year)})
        Exports_Country_Dprice=Weight_Exports_Country.loc[:,['Country','Weight_DPrice']].rename(columns={'Weight_DPrice':"DPrice_Exp_"+str(year)})


    #    Imports_Custom_Dprice=Weight_Imports_Custom.loc[:,['Custom','Weight_DPrice']].rename(columns={'Weight_DPrice':"DPrice_Imp_"+str(year)})
    #    Exports_Custom_Dprice=Weight_Exports_Custom.loc[:,['Custom','Weight_DPrice']].rename(columns={'Weight_DPrice':"DPrice_Exp_"+str(year)})


        Imports_TCode_Dprice=Weight_Imports_TCode.loc[:,['Tcode','Tdesc','Weight_DPrice']].rename(columns={'Weight_DPrice':"DPrice_Imp_"+str(year)})
        Exports_TCode_Dprice=Weight_Exports_TCode.loc[:,['Tcode','Tdesc','Weight_DPrice']].rename(columns={'Weight_DPrice':"DPrice_Exp_"+str(year)})


        Imports_Country_TCode_DPrice=Weight_Imports_Country_Tcode.loc[:,['Country','Tcode','Tdesc','Weight_DPrice']].rename(columns={'Weight_DPrice':"DPrice_Imp_"+str(year)})
        Exports_Country_TCode_DPrice=Weight_Exports_Country_Tcode.loc[:,['Country','Tcode','Tdesc','Weight_DPrice']].rename(columns={'Weight_DPrice':"DPrice_Exp_"+str(year)})   


    else:
         Imports_Country_Rprice=Imports_Country_Rprice.merge(Weight_Imports_Country.loc[:,['Country','Weight_RPrice']].rename(columns={'Weight_RPrice':"RPrice_Imp_"+str(year)}),how='outer',on='Country')
         Exports_Country_Rprice=Exports_Country_Rprice.merge(Weight_Exports_Country.loc[:,['Country','Weight_RPrice']].rename(columns={'Weight_RPrice':"RPrice_Exp_"+str(year)}),how='outer',on='Country')

    #    Imports_Custom_Rprice=Imports_Custom_Rprice.merge(Weight_Imports_Custom.loc[:,['Custom','Weight_RPrice']].rename(columns={'Weight_RPrice':"RPrice_Imp_"+str(year)}),how='outer',on='Custom')
    #    Exports_Custom_Rprice=Exports_Custom_Rprice.merge(Weight_Exports_Custom.loc[:,['Custom','Weight_RPrice']].rename(columns={'Weight_RPrice':"RPrice_Exp_"+str(year)}),how='outer',on='Custom')

    #    Imports_TCode_Rprice=Imports_TCode_Rprice.merge(Weight_Imports_TCode.loc[:,['Tcode','Tdesc','Weight_RPrice']].rename(columns={'Weight_RPrice':"RPrice_Imp_"+str(year)}),how='outer',on=['Tcode','Tdesc'])
    #    Exports_TCode_Rprice=Exports_TCode_Rprice.merge(Weight_Exports_TCode.loc[:,['Tcode','Tdesc','Weight_RPrice']].rename(columns={'Weight_RPrice':"RPrice_Exp_"+str(year)}),how='outer',on=['Tcode','Tdesc'])

    #    Imports_Country_TCode_RPrice=Imports_Country_TCode_RPrice.merge(Weight_Imports_Country_Tcode.loc[:,['Country','Tcode','Tdesc','Weight_RPrice']].rename(columns={'Weight_RPrice':"RPrice_Imp_"+str(year)}), on=['Country','Tcode','Tdesc'], how='outer')
    #    Exports_Country_TCode_RPrice=Exports_Country_TCode_RPrice.merge(Weight_Exports_Country_Tcode.loc[:,['Country','Tcode','Tdesc','Weight_RPrice']].rename(columns={'Weight_RPrice':"RPrice_Exp_"+str(year)}), on=['Country','Tcode','Tdesc'], how='outer')   





         Imports_Country_Dprice=Imports_Country_Dprice.merge(Weight_Imports_Country.loc[:,['Country','Weight_DPrice']].rename(columns={'Weight_DPrice':"DPrice_Imp_"+str(year)}),how='outer',on='Country')
         Exports_Country_Dprice=Exports_Country_Dprice.merge(Weight_Exports_Country.loc[:,['Country','Weight_DPrice']].rename(columns={'Weight_DPrice':"DPrice_Exp_"+str(year)}),how='outer',on='Country')

    #    Imports_Custom_Dprice=Imports_Custom_Dprice.merge(Weight_Imports_Custom.loc[:,['Custom','Weight_DPrice']].rename(columns={'Weight_DPrice':"DPrice_Imp_"+str(year)}),how='outer',on='Custom')
    #    Exports_Custom_Dprice=Exports_Custom_Dprice.merge(Weight_Exports_Custom.loc[:,['Custom','Weight_DPrice']].rename(columns={'Weight_DPrice':"DPrice_Exp_"+str(year)}),how='outer',on='Custom')

      #  Imports_TCode_Dprice=Imports_TCode_Dprice.merge(Weight_Imports_TCode.loc[:,['Tcode','Tdesc','Weight_DPrice']].rename(columns={'Weight_DPrice':"DPrice_Imp_"+str(year)}),how='outer',on=['Tcode','Tdesc'])
      #  Exports_TCode_Dprice=Exports_TCode_Dprice.merge(Weight_Exports_TCode.loc[:,['Tcode','Tdesc','Weight_DPrice']].rename(columns={'Weight_DPrice':"DPrice_Exp_"+str(year)}),how='outer',on=['Tcode','Tdesc'])

    
         Imports_Country_TCode_DPrice=Imports_Country_TCode_DPrice.merge(Weight_Imports_Country_Tcode.loc[:,['Country','Tcode','Tdesc','Weight_DPrice']].rename(columns={'Weight_DPrice':"DPrice_Imp_"+str(year)}), on=['Country','Tcode','Tdesc'], how='outer')
         Exports_Country_TCode_DPrice=Exports_Country_TCode_DPrice.merge(Weight_Exports_Country_Tcode.loc[:,['Country','Tcode','Tdesc','Weight_DPrice']].rename(columns={'Weight_DPrice':"DPrice_Exp_"+str(year)}), on=['Country','Tcode','Tdesc'], how='outer')   

    print(f"Year: {year}")

```
